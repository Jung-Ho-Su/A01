# 컨벌루션 및 맥스풀링

```python
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

(x_train, t_train), (x_test, t_test) = mnist.load_data()
```

```python
x_train.shape, t_train.shape, x_test.shape, t_test.shape
```

```python
# 데이터 구성을 확인해봣으니 데이터를 출력해보자
plt.imshow(x_train[2], cmap='gray')
```

```python
# 학습했다 가정하고 필터를 정의해보자
# 원래 이 필터 값을 학습을 통해서 찾아야하는데 지금은 그냥 가정하고 해보자
# 여기서 필터 값은 [] 안의 숫자 값인 1. 0. -1. 이런 친구들이다
import numpy as np

horizontal_filter = np.array([[1., 1., 1.], 
                             [0., 0., 0.],
                             [-1., -1., -1.]])
vertical_filter = np.array([[1., 0., -1.],
                           [1., 0., -1.],
                           [1., 0., -1.]])
```

```python
test_image = x_train[2]
original_image_size = test_image.shape[0] # test_imageshape : (28, 28)
print("original_image_size = ", original_image_size)
conv_output_size = int((original_image_size - 3)/1 + 1)
print("output size = ", conv_output_size)
```

```python
filter_size = 3

def get_conv_image(filter):
    filtered_image = np.zeros((conv_output_size, conv_output_size))# 피쳐맵을 저장ㅇ하기 위한 배열
    for i in range(conv_output_size):
        for j in range(conv_output_size):
            indice_image = test_image[i:(i + filter_size), j:(j + filter_size)] * filter
            indice_sum = np.sum(indice_image)
            if (indice_sum > 255):
                indice_sum = 255

            filtered_image[i, j] = indice_sum # 피처맵에 저장하는 과정 

    return filtered_image # 이후 26행 26열 짜리의 피쳐맵이 만들어짐
```

```python
vertical_filtered_image = get_conv_image(vertical_filter)
horizontal_filtered_image = get_conv_image(horizontal_filter)
```

```python
plt.subplot(1, 2, 1)
plt.title('vertical')
plt.imshow(vertical_filtered_image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('horizontal')
plt.imshow(horizontal_filtered_image, cmap='gray')
plt.show()
```

```python
# 수평 수직 각각 경계선이 잘 나타나는 것을 확인했으니 합쳐보자!
```

```python
sobel_image = np.sqrt(np.square(horizontal_filtered_image) + np.square(vertical_filtered_image))
```

```python
plt.imshow(sobel_image, cmap='gray')
```

```python
# 원본 이미지 가지고 풀링을 적용하는 예시

import numpy as np

image = x_train[2]
image_x = image.shape[0]
image_y = image.shape[1]

new_image_x = int(image_x / 2)
new_image_y = int(image_y / 2)

pooled_image = np.zeros((new_image_x, new_image_y))
for x in range(0, image_x, 2):                                                          # 여기서 풀링 연산 풀 사이즈는 (2, 2) 사이즈
    for y in range(0, image_y, 2):
        pooled_image[int(x/2), int(y/2)] = np.max(image[x:x + 2, y:y + 2])              # 맥스 풀링 

plt.imshow(pooled_image, cmap='gray')
```

```python
# 원본 이미지 가지고 풀링을 적용하는 예시
# image = sobel_image 로 바꾼 것

import numpy as np

image = sobel_image
image_x = image.shape[0]
image_y = image.shape[1]

new_image_x = int(image_x / 2)
new_image_y = int(image_y / 2)

pooled_image = np.zeros((new_image_x, new_image_y))
for x in range(0, image_x, 2):                                                          # 여기서 풀링 연산 풀 사이즈는 (2, 2) 사이즈
    for y in range(0, image_y, 2):
        pooled_image[int(x/2), int(y/2)] = np.max(image[x:x + 2, y:y + 2])              # 맥스 풀링 

plt.imshow(pooled_image, cmap='gray')
```

# Minist 데이터 분류

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for index in range(25):
    plt.subplot(5, 5, index+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[index], cmap='gray')
    plt.xlabel(str(y_train[index]))
plt.show()
```

#### Conv2D(16, 3 or (3,3), padding='valid', input_shape=(28, 28, 1), activation='relu')
- 첫번째 인자 : 컨볼루션 필터의 수
- 두번째 인자 : 컨볼루션 커널의 (행, 열) 크기
- padding : 경계 처리 방법 정의.
    - 'valid' : 유효한 영역만 출력.출력 이미지 사이즈는 입력 사이즈보다 작아진다.
    - 'same' : 출력 이미지 사이즈가 입력 이미지 사이즈와 동일하다.
- input_shape : 샘플 수를 제외한 입력 형태 정의. 모델에서 첫 레이어일 때만 정의한다.
    - (행, 열, 채널 수)로 정의한다. 흑백영상인 경우에는 채널이 1이고, 컬러(RGB)영상인 경우에는 채널을 3으로 설정한다.
- activation : 활성화 함수를 설정한다.
    - 'linear' : 입력뉴런과 가중치로 계산된 결과값이 그대로 출력으로 나온다.default=None
    - 'relu' : rectifier 함수, 은익층에 주로 쓰인다.
    - 'sigmoid' : 시그모이드 함수, 이진 분류 문제에서 출력층에 주로 쓰인다.
    - 'softmax' : 소프트맥스 함수, 다중 클래스 분류 문제에서 출력층에 주로 쓰인다.

```python
# 모델 설계

model = tf.keras.models.Sequential([
    tf.keras.layers.Reshape((28, 28, 1)), # 행 열 채널 // # 높이 너비 채널
    tf.keras.layers.Conv2D(16, 3, activation='relu'), # 2차원 컨블루션 연산 (16, 3, activation='relu')
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5)
```

```python
model.evaluate(x_test, y_test, verbose=2)
```

```python
model.summary()
```

### Keras API ( Sequential vs Functional )
Tensorflow에서 keras api 를 활용하는 방법이 크게 두 가지.
    
    - Sequential은 layer의 구조가 순차적으로 있을 때 사용이 편리.
    
    - Functional API를 활용한다면 복잡한 구조를 설계할 수 있음.
    
        - 예를 들어서 input, output을 여러개 관리 하거나, layer 사이사이에 연결을 해줄 때.

# Functional API 방식을 이용한 예제

```python
# 함수 호출하듯이 모델을 차곡차곡 연결시켜주는 functionalapi 방식

inputs = keras.Input(shape=(28, 28))
x = layers.Reshape((28, 28, 1))(inputs) # 함수의 파라매터
x = layers.Conv2D(16, 3, activation='relu')(x) # x를 가지고 와서 컨블루션 연산할거임
x = layers.Flatten()(x)                        # x를 가지고 와서 1차원으로 변경
x = layers.Dense(128)(x)                       # x를 가자ㅣ고 와서 은닉층 만듦
x = layers.Dense(10, activation='softmax')(x)  # x를 가지고 와서 출력층 만듦
outputs = x

model = keras.Model(inputs, outputs)           # 함수를 쭉 정의하고 이제 모델 설계를 함 (입력, 출력 )구조로 만듦 // 여러개의 파라매터가 들어가면 리스트 형태로 넣어주면 된다.
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5)
model.evaluate(x_test, y_test, verbose=0)
```

```python
### 현재 컨블루션 연산만 했으니 이제 풀링을 한번 적용해보자
```

```python
import numpy as np 
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
```

```python
# 데이터 셋을 변형시켜주는 부분
x_train = np.reshape(x_train, (60000, 28, 28, 1))
x_test = np.reshape(x_test, (10000, 28, 28, 1))
```

```python
# 모델 설계
model = Sequential()
model.add(Conv2D(16, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5)
model.evaluate(x_test, y_test, verbose=2)
```

```python
# 모델 설계
model = Sequential()
model.add(Conv2D(16, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(16, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5)
model.evaluate(x_test, y_test, verbose=2)
```

```python
model = Sequential()
model.add(Conv2D(16, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1), padding="same"))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(16, kernel_size=(3, 3), activation='relu', padding="same"))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()

# 패딩을 적용하게 되면 패치전 사이즈가 원 사이즈로 유지하게 된다.

# 맥스 풀링을 하면 반으로 줄어들고
# 컨불루션 연산을하고 반으로
# 다시 맥스 풀링하니 반으로 줄어들게 된다.
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5)
model.evaluate(x_test, y_test, verbose=2)
```

# 이미지 분류 예제 - cifar10

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
```

```python
cifar10 =datasets.cifar10
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

```python
print("Train samples:", train_images.shape, train_labels.shape)
print("Test samples:", test_images.shape, test_labels.shape)
```

```python
train_images[0]
```

```python
train_labels[:10]
```

```python
# 클래스 네임이 따로 정리 되지 않았기 때문에 정리해주었다.
class_names = ['airplane', 'automobile', 'bird', ' cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i][0]])
plt.show()
```

```python
# 이미지 작업에 대한 스케일링 작업

train_images = train_images/255.0
test_images = test_images/255.0
```

```python
model = Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D(2))                                                 # 풀링을 적용할 때 사이즈 = 2

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))                           # 마지막에는 컨블루션 연산만 했ㅇ다
model.add(layers.Flatten())

model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python

model.fit(train_images, train_labels, epochs=10)
```

```python
# 이거는 컬리이기 때문에 흑백보다는 경계선이 잘 되지 않는 경우가 생긴다. 
# 그렇기 때문에 openCV에서 배운 이미지 전처리 작업을 해줘서 경계선을 정확하게 해줘야한다.
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Test loss:', test_loss)
print('Test accuracy:', test_acc)
```
