### [문제] 타이타닉 데이터(./titanic/train.csv)를 사용하여 딥러닝 모델을 구현하여 마지막에 정확도를 출력하는 코드를 작성하시오.

제출형식: 홍길동_0522_정확도.ipynb

<처리조건>   
1. 텐서플로우가 제공하는 케라스를 사용한다.
2. 딥러닝 모델로 구현한다.
3. 정확도를 최대한 높일 수 있도록 데이터에 대한 전처리를 수행한다.
4. 마지막에 테스트 데이터에 대한 정확도를 출력한다.
5. 훈련 데이터와 테스트 데이터는 ./titanic/train.csv를 사용하고 8:2로 한다.

```python
# 결측치 어떻게 처리 # 피쳐 선택 여부 판단 # 문자 데이터인 경우 인코딩 할 때 어떤 방식으로 할지 체크

import pandas as pd
import tensorflow as tf
from tensorflow import keras
import numpy as np

data = pd.read_csv("./titanic_train.csv")
print(data.isnull().sum())

# 독립, 종속 변수 지정
xdata = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']] # 7
tdata = pd.DataFrame(data['Survived'])
# 독립 : PassengerId Pclass Name Sex	Age	SibSp	Parch	Ticket	Fare	Cabin	Embarked
# 종속 : Survived
# PassengerId : 의미없음
# Name : 의미 없음
# Ticket : 의미 없음
# Cabin : 결측치가 너무 많음
```

```python
# 전처리
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

# Sex
encoder = LabelEncoder() # 라벨
xdata['Sex'] = encoder.fit_transform(xdata['Sex'])

# Embarked # 라벨
xdata['Embarked'] = xdata['Embarked'].fillna(xdata['Embarked'].mode()[0])
xdata['Embarked'] = encoder.fit_transform(xdata['Embarked'])

# Age, Fare # 민 맥스
xdata['Age'] = xdata['Age'].fillna(xdata['Age'].median())
scaler = MinMaxScaler()
xdata[['Age', 'Fare']] =scaler.fit_transform(xdata[['Age', 'Fare']])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(xdata, tdata, test_size=0.2, random_state=42)

model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(7,))) 
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))   # 은닉층
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))  # 출력 # 이진분류 # 시그모이드
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=64, verbose=1)
```

```python
result_0522 = model.evaluate(X_test, y_test)
result_0522
```
