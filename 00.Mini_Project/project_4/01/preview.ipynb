```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 한글 폰트 설정 (Windows 환경의 맑은 고딕)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 불러오기
file_name = '2023년도 전시회 요약.csv'

try:
    # 데이터 읽기
    df = pd.read_csv(file_name)

    # 3. '1월'부터 '12월'까지의 컬럼 리스트 생성
    month_cols = [f'{i}월' for i in range(1, 13)]
    
    # 4. 각 월 컬럼의 데이터 합계 구하기 (2행의 숫자 데이터 합산)
    # 만약 데이터에 문자가 섞여 있을 경우를 대비해 숫자형으로 변환합니다.
    monthly_data = df[month_cols].apply(pd.to_numeric, errors='coerce').sum()

    # 5. 막대그래프 시각화
    plt.figure(figsize=(12, 7))
    
    # x축은 '1월', '2월'..., y축은 합산된 전시회 수
    bars = plt.bar(monthly_data.index, monthly_data.values, color='dodgerblue', edgecolor='black', alpha=0.8)

    # 그래프 스타일 설정
    plt.title('2023년 인천 송도 컨벤션센터 월별 전시회 개최 현황', fontsize=18, pad=20, fontweight='bold')
    plt.xlabel('개최 월', fontsize=13)
    plt.ylabel('전시회 개최 건수', fontsize=13)
    plt.xticks(fontsize=11)
    plt.grid(axis='y', linestyle='--', alpha=0.5)

    # 막대 위에 숫자 표시
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height + 0.2, f'{int(height)}', 
                 ha='center', va='bottom', fontsize=11, fontweight='bold')

    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print(f"❌ 오류: '{file_name}' 파일을 찾을 수 없습니다.")
except KeyError as e:
    print(f"❌ 오류: 컬럼명을 찾을 수 없습니다. ({e}) 파일 내 컬럼명이 '1월', '2월' 형식인지 확인해주세요.")
except Exception as e:
    print(f"❌ 기타 오류 발생: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 불러오기
file_name = 'top5_전체통합_23-25.csv'

try:
    # 데이터 로드
    df = pd.read_csv(file_name)

    # 3. 2023년 월별 데이터 위치 설정 (Unnamed: 6 ~ Unnamed: 17 컬럼이 1월~12월에 해당)
    # 데이터 프레임의 컬럼 구조에 따라 인덱스를 슬라이싱합니다.
    # 0:시도, 1:군구, 2:관광지, 3:내/외국인, 4:총계, 5:인원계(2023), 6~17:1~12월
    data_2023 = df.iloc[:, 6:18].apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',', ''), errors='coerce')).fillna(0)
    
    # 4. 내국인/외국인별 합산
    # '내/외국인' 컬럼(인덱스 3)을 기준으로 필터링
    korean_monthly = data_2023[df.iloc[:, 3] == '내국인'].sum()
    foreign_monthly = data_2023[df.iloc[:, 3] == '외국인'].sum()
    total_monthly = korean_monthly + foreign_monthly

    # 5. 그래프 시각화
    x_labels = [f'{i}월' for i in range(1, 13)]
    x = np.arange(len(x_labels))
    width = 0.35  # 막대 너비

    plt.figure(figsize=(15, 8))

    # [막대그래프] 1월 위치에 내국인, 외국인 막대 2개를 나란히 배치
    plt.bar(x - width/2, korean_monthly, width, label='내국인', color='#3498db', alpha=0.7, edgecolor='navy')
    plt.bar(x + width/2, foreign_monthly, width, label='외국인', color='#e74c3c', alpha=0.7, edgecolor='darkred')

    # [꺾은선그래프] 내국인+외국인 합산 추이 (막대들의 합)
    plt.plot(x, total_monthly, color='#2c3e50', marker='o', linestyle='-', linewidth=3, markersize=8, label='전체 합계 추이')

    # 그래프 세부 설정
    plt.title('2023년 월별 내/외국인 입장객 현황 및 합계 추이', fontsize=18, pad=20, fontweight='bold')
    plt.xlabel('2023년도', fontsize=12)
    plt.ylabel('입장객 수', fontsize=12)
    plt.xticks(x, x_labels)
    plt.legend(fontsize=11, loc='upper right')
    plt.grid(axis='y', linestyle='--', alpha=0.4)

    # 합계 수치 강조 (꺾은선 포인트 위에 숫자 표시)
    for i, txt in enumerate(total_monthly):
        plt.annotate(f'{int(txt):,}', (x[i], total_monthly[i]), textcoords="offset points", xytext=(0,10), ha='center', fontweight='bold', color='#2c3e50')

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"❌ 오류 발생: {e}")
    print("💡 CSV 파일의 컬럼 구조가 예상과 다를 수 있습니다. iloc 슬라이싱 범위를 확인해주세요.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. 한글 폰트 설정 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 불러오기
file_name = 'Top5_관광지_상세내역(중복제거).csv'

try:
    # 데이터 로드
    df = pd.read_csv(file_name)

    # 3. 2023년 월별 데이터 컬럼 (6번째 컬럼부터 17번째 컬럼까지가 1월~12월)
    # 데이터 전처리: 콤마 제거 및 숫자형 변환
    data_2023 = df.iloc[:, 6:18].apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',', ''), errors='coerce')).fillna(0)
    
    # 4. 내국인 / 외국인 행 분리 후 월별 합계 계산
    # '내/외국인' 컬럼(인덱스 3)을 기준으로 필터링
    korean_total_monthly = data_2023[df.iloc[:, 3] == '내국인'].sum()
    foreign_total_monthly = data_2023[df.iloc[:, 3] == '외국인'].sum()
    
    # 두 값을 합친 전체 추이 데이터
    overall_total_monthly = korean_total_monthly + foreign_total_monthly

    # 5. 그래프 시각화
    x_labels = [f'{i}월' for i in range(1, 13)]
    x = np.arange(len(x_labels))
    width = 0.35

    fig, ax1 = plt.subplots(figsize=(15, 8))

    # [막대그래프] 내국인(왼쪽), 외국인(오른쪽) 막대 배치
    bar1 = ax1.bar(x - width/2, korean_total_monthly, width, label='내국인 합계', color='#3498db', alpha=0.8)
    bar2 = ax1.bar(x + width/2, foreign_total_monthly, width, label='외국인 합계', color='#e74c3c', alpha=0.8)

    # [꺾은선그래프] 내국인+외국인 통합 추이 (동일 축 사용)
    line1 = ax1.plot(x, overall_total_monthly, color='#2c3e50', marker='s', linestyle='-', linewidth=3, markersize=8, label='전체 합계 추이')

    # 그래프 스타일링
    ax1.set_title('2023년 Top 5 관광지 월별 입장객 현황 및 통합 추이', fontsize=18, pad=20, fontweight='bold')
    ax1.set_xlabel('2023년도', fontsize=12)
    ax1.set_ylabel('입장객 수 (명)', fontsize=12)
    ax1.set_xticks(x)
    ax1.set_xticklabels(x_labels)
    ax1.legend(fontsize=11)
    ax1.grid(axis='y', linestyle='--', alpha=0.4)

    # 꺾은선 그래프 위에 합계 수치 표시
    for i, val in enumerate(overall_total_monthly):
        ax1.text(i, val + (val * 0.02), f'{int(val):,}', ha='center', va='bottom', fontsize=10, fontweight='bold', color='#2c3e50')

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"❌ 오류 발생: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # [수정] import np를 import numpy as np로 변경

# 1. 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 이름 설정
file_tour = 'Top5_관광지_상세내역(중복제거).csv'
file_exhibition = '2023년도 전시회 요약.csv'

try:
    # --- 데이터 1: 관광지 방문객 데이터 정리 ---
    df_tour = pd.read_csv(file_tour)
    data_2023_tour = df_tour.iloc[:, 6:18].apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',', ''), errors='coerce')).fillna(0)
    tour_monthly = data_2023_tour[df_tour.iloc[:, 3] == '합계'].sum()

    # --- 데이터 2: 전시회 개최 데이터 정리 ---
    df_exh = pd.read_csv(file_exhibition)
    exh_monthly = df_exh.iloc[0, 0:12].apply(pd.to_numeric).fillna(0)

    # 3. 시각화 설정
    months = [f'{i}월' for i in range(1, 13)]
    x = np.arange(len(months))

    # 그래프 생성
    fig, ax1 = plt.subplots(figsize=(16, 9), constrained_layout=True)

    # 4. 그래프 중앙 상단 제목 (pad로 간격 조정)
    plt.title('2023년 월별 관광지 방문객 및 전시회 개최 현황 비교', fontsize=22, fontweight='bold', pad=30)

    # [좌측 축] 관광지 방문객 수 추이 (파란색)
    line1 = ax1.plot(x, tour_monthly, color='#3498db', marker='o', linestyle='-', 
                     linewidth=4, markersize=10, label='관광지 방문객 수 (명)')
    ax1.set_ylabel('방문객 수 (명)', fontsize=14, fontweight='bold', color='#3498db')
    ax1.tick_params(axis='y', labelcolor='#3498db')
    ax1.set_xlabel('2023년도', fontsize=13)

    # [우측 축] 전시회 개최 건수 추이 (빨간색)
    ax2 = ax1.twinx()
    line2 = ax2.plot(x, exh_monthly, color='#e74c3c', marker='s', linestyle='--', 
                     linewidth=4, markersize=10, label='전시회 개최 건수 (건)')
    ax2.set_ylabel('전시회 개최 건수 (건)', fontsize=14, fontweight='bold', color='#e74c3c', rotation=270, labelpad=25)
    ax2.tick_params(axis='y', labelcolor='#e74c3c')

    # 수치 표시
    for i in range(len(months)):
        # 방문객 수
        ax1.text(i, tour_monthly[i] + (tour_monthly.max() * 0.02), f'{int(tour_monthly[i]):,}', 
                 ha='center', va='bottom', fontsize=11, fontweight='bold', color='#3498db')
        # 전시회 건수
        ax2.text(i, exh_monthly[i] - (exh_monthly.max() * 0.05), f'{int(exh_monthly[i])}건', 
                 ha='center', va='top', fontsize=11, fontweight='bold', color='#e74c3c')

    # X축 설정
    ax1.set_xticks(x)
    ax1.set_xticklabels(months, fontsize=11)
    
    # 범례 통합 (좌측 상단)
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc='upper left', fontsize=13, frameon=True, shadow=True)

    # 그리드 및 범위 최적화
    ax1.grid(axis='y', linestyle=':', alpha=0.6)
    ax1.set_ylim(0, tour_monthly.max() * 1.2)
    ax2.set_ylim(0, exh_monthly.max() * 1.3)

    plt.show()

except Exception as e:
    # 혹시라도 다른 오류가 나면 내용을 확인할 수 있게 출력만 살려둡니다.
    print(f"데이터를 불러오는 중 문제가 발생했습니다: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. 한글 폰트 설정 (Windows: Malgun Gothic, Mac: AppleGothic)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 불러오기
file_name = 'Top5_관광지_상세내역(중복제거).csv'

try:
    df = pd.read_csv(file_name)

    # 3. 2025년 월별 데이터 추출 및 숫자 변환
    # iloc[:, 32:44]는 2025년 01월부터 12월까지의 데이터 컬럼에 해당합니다.
    data_2025 = df.iloc[:, 32:44].apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',', ''), errors='coerce')).fillna(0)
    
    # 4. '합계' 행만 추출하여 월별 방문객 수 총합 계산
    total_monthly_2025 = data_2025[df.iloc[:, 3] == '합계'].sum()

    # 5. 그래프 시각화 설정
    x_labels = [f'{i}월' for i in range(1, 13)]
    x = np.arange(len(x_labels))

    fig, ax1 = plt.subplots(figsize=(15, 8))

    # [막대그래프] 방문객 합계 (단일 막대)
    # x 좌표 중앙에 배치하여 꺾은선 점과 일치시킵니다.
    ax1.bar(x, total_monthly_2025, width=0.6, color='#3498db', alpha=0.7, label='방문객 수', zorder=2)

    # [꺾은선그래프] 방문객 합계 추이
    # 막대 정중앙에 점이 오도록 x 좌표를 그대로 사용합니다.
    ax1.plot(x, total_monthly_2025, color='#2c3e50', marker='o', linestyle='-', 
             linewidth=3, markersize=10, label='방문객 수 (추이)', zorder=3)

    # 그래프 세부 설정
    ax1.set_title('2025년 Top 5 관광지 월별 통합 방문객 현황', fontsize=18, pad=20, fontweight='bold')
    ax1.set_xlabel('2025년도', fontsize=12)
    ax1.set_ylabel('방문객 수 (명)', fontsize=12)
    
    # X축 눈금을 막대와 점의 중앙에 맞춤
    ax1.set_xticks(x)
    ax1.set_xticklabels(x_labels)
    
    # 범례 설정: 우측 상단 박스 (그림자 효과 포함)
    ax1.legend(loc='upper right', fontsize=11, frameon=True, shadow=True)
    
    # 그리드 설정 (막대 뒤로 배치)
    ax1.grid(axis='y', linestyle='--', alpha=0.4, zorder=1)

    # 수치 표시 (꺾은선 점 바로 위에 콤마 포함하여 표시)
    max_val = total_monthly_2025.max()
    for i, val in enumerate(total_monthly_2025):
        if val > 0: # 데이터가 있는 경우만 표시
            ax1.text(i, val + (max_val * 0.02), f'{int(val):,}', 
                     ha='center', va='bottom', fontsize=10, fontweight='bold', color='#2c3e50')

    # 좌우 여백 조정 (막대가 테두리에 붙지 않도록)
    ax1.set_xlim(-0.8, len(x)-0.2)

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"❌ 오류 발생: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import platform

# 1. 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin':  # Mac
    plt.rcParams['font.family'] = 'AppleGothic'
else:
    plt.rcParams['font.family'] = 'NanumGothic'

plt.rcParams['axes.unicode_minus'] = False 

# 2. 관광소비 데이터 로드 (2023년 데이터 파일 5개 통합)
# 파일명이 '20260408...' 형태이므로 해당 패턴으로 읽어옵니다.
tour_files = glob.glob('20260408*_관광소비 추이.csv')

if not tour_files:
    print("관광소비 추이 CSV 파일을 찾을 수 없습니다. 파일명을 확인해주세요.")
else:
    dfs = []
    for f in tour_files:
        temp_df = pd.read_csv(f)
        dfs.append(temp_df)
    
    df_tour = pd.concat(dfs, ignore_index=True)

    # 2023년 데이터 필터링
    df_tour['기준년월'] = df_tour['기준년월'].astype(str)
    df_2023 = df_tour[df_tour['기준년월'].str.startswith('2023')].copy()

    # 카테고리 매핑 함수
    def map_category(x):
        if x in ['호텔', '기타숙박']:
            return '숙박'
        elif x in ['대형쇼핑몰', '기타관광쇼핑', '일반외식업', '제과음료업']:
            return '상권'
        return '기타'

    df_2023['분류'] = df_2023['중분류'].apply(map_category)

    # 월별/분류별 소비액 합산
    tour_summary = df_2023[df_2023['분류'] != '기타'].groupby(['기준년월', '분류'])['소비액(천원)'].sum().unstack()
    
    # 데이터 정렬 및 X축 라벨 설정
    tour_summary = tour_summary.sort_index()
    months = [f"{i}월" for i in range(1, 13)]

    # 3. 전시회 데이터 로드 (2023년도 전시회 요약.csv)
    df_exhib = pd.read_csv('2023년도 전시회 요약.csv')
    # 파일의 첫 번째 행에서 1~12월 개최 건수(0~11번 컬럼) 추출
    exhib_counts = df_exhib.iloc[0, 0:12].astype(int).values

    # --- 그래프 시각화 함수 ---
    def draw_comparison_graph(data_series, exhib_data, title, y1_label, color):
        plt.figure(figsize=(12, 6))
        ax1 = plt.gca()
        
        # 좌측 Y축: 소비액
        ax1.plot(months, data_series, marker='o', color=color, linewidth=2, label=y1_label)
        ax1.set_xlabel('월 (2023년)')
        ax1.set_ylabel(f'{y1_label} (천원)', color=color)
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.grid(True, axis='y', linestyle=':', alpha=0.7)

        # 우측 Y축: 전시회 개수
        ax2 = ax1.twinx()
        ax2.plot(months, exhib_data, linestyle='--', marker='^', color='red', linewidth=2, label='전시회 개수')
        ax2.set_ylabel('전시회 개수 (건)', color='red')
        ax2.tick_params(axis='y', labelcolor='red')

        plt.title(title)
        
        # 범례 통합
        lines1, labels1 = ax1.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
        
        plt.tight_layout()
        plt.show()

    # 그래프 1: 상권 소비
    draw_comparison_graph(tour_summary['상권'], exhib_counts, 
                          '2023년 송도 전시회 개최 현황 및 상권 소비 추이 비교', 
                          '상권 소비액', 'blue')

    # 그래프 2: 숙박 소비
    draw_comparison_graph(tour_summary['숙박'], exhib_counts, 
                          '2023년 송도 전시회 개최 현황 및 숙박 소비 추이 비교', 
                          '숙박 소비액', 'green')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import platform

# 1. 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin':  # Mac
    plt.rcParams['font.family'] = 'AppleGothic'
else:
    plt.rcParams['font.family'] = 'NanumGothic'

plt.rcParams['axes.unicode_minus'] = False 

# 2. 관광소비 데이터 로드 (2025년 포함 데이터 파일 5개 통합)
tour_files = glob.glob('20260408*_관광소비 추이.csv')

if not tour_files:
    print("관광소비 추이 CSV 파일을 찾을 수 없습니다.")
else:
    dfs = []
    for f in tour_files:
        temp_df = pd.read_csv(f)
        dfs.append(temp_df)
    
    df_tour = pd.concat(dfs, ignore_index=True)

    # 2025년 데이터 필터링
    df_tour['기준년월'] = df_tour['기준년월'].astype(str)
    df_2025 = df_tour[df_tour['기준년월'].str.startswith('2025')].copy()

    # 카테고리 매핑 함수
    def map_category(x):
        if x in ['호텔', '기타숙박']:
            return '숙박'
        elif x in ['대형쇼핑몰', '기타관광쇼핑', '일반외식업', '제과음료업']:
            return '상권'
        return '기타'

    df_2025['분류'] = df_2025['중분류'].apply(map_category)

    # 월별/분류별 소비액 합산
    tour_summary = df_2025[df_2025['분류'] != '기타'].groupby(['기준년월', '분류'])['소비액(천원)'].sum().unstack()
    
    # 데이터 정렬 및 X축 라벨 설정
    tour_summary = tour_summary.sort_index()
    months = [f"{i}월" for i in range(1, 13)]

    # 3. 전시회 데이터 로드 (2025년도 전시회 요약.csv)
    df_exhib = pd.read_csv('2025년도 전시회 요약.csv')
    # 파일의 첫 번째 행에서 1~12월 개최 건수 추출
    exhib_counts = df_exhib.iloc[0, 0:12].fillna(0).astype(int).values

    # --- 그래프 시각화 함수 ---
    def draw_comparison_graph(data_series, exhib_data, title, y1_label, color):
        plt.figure(figsize=(12, 6))
        ax1 = plt.gca()
        
        # 좌측 Y축: 소비액
        ax1.plot(months, data_series, marker='o', color=color, linewidth=2, label=y1_label)
        ax1.set_xlabel('월 (2025년)')
        ax1.set_ylabel(f'{y1_label} (천원)', color=color)
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.grid(True, axis='y', linestyle=':', alpha=0.7)

        # 우측 Y축: 전시회 개수
        ax2 = ax1.twinx()
        ax2.plot(months, exhib_data, linestyle='--', marker='^', color='red', linewidth=2, label='전시회 개수')
        ax2.set_ylabel('전시회 개수 (건)', color='red')
        ax2.tick_params(axis='y', labelcolor='red')

        plt.title(title)
        
        # 범례 통합
        lines1, labels1 = ax1.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
        
        plt.tight_layout()
        plt.show()

    # --- 최종 결과 그래프 출력 ---
    
    # 그래프 1: 2025년 상권 소비 비교
    if '상권' in tour_summary.columns:
        draw_comparison_graph(tour_summary['상권'], exhib_counts, 
                              '2025년 송도 전시회 개최 현황 및 상권 소비 추이 비교', 
                              '상권 소비액', 'blue')
    else:
        print("2025년 상권 소비 데이터가 부족합니다.")

    # 그래프 2: 2025년 숙박 소비 비교
    if '숙박' in tour_summary.columns:
        draw_comparison_graph(tour_summary['숙박'], exhib_counts, 
                              '2025년 송도 전시회 개최 현황 및 숙박 소비 추이 비교', 
                              '숙박 소비액', 'green')
    else:
        print("2025년 숙박 소비 데이터가 부족합니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import platform

# 1. 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
else:
    plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 경로 설정
file_songdo = '송도 연별 개최건수.xlsx'
file_biz = '전시사업자지역_업종별_사업체수_전시관련종사자수_2018~2023.xlsx'

try:
    # --- 데이터 1: 송도 개최 건수 처리 ---
    df_s = pd.read_excel(file_songdo)
    # 가로 데이터를 세로로 변환
    df_s_trend = df_s.set_index(df_s.columns[0]).T.reset_index()
    df_s_trend.columns = ['연도', '개최건수']
    df_s_trend['연도'] = df_s_trend['연도'].astype(int)
    # 2018~2023년 데이터만 추출
    df_s_filtered = df_s_trend[(df_s_trend['연도'] >= 2018) & (df_s_trend['연도'] <= 2023)]

    # --- 데이터 2: 인천 사업체 수 처리 ---
    # 헤더가 여러 줄인 경우를 고려하여 읽기
    df_b = pd.read_excel(file_biz, header=[0, 1, 2])
    
    # '인천' 지역 행 찾기 (두 번째 컬럼 '특성별(2)'에서 검색)
    incheon_row = df_b[df_b.iloc[:, 1].str.contains('인천', na=False)].iloc[0]
    
    years = [2018, 2019, 2020, 2021, 2022, 2023]
    biz_counts = []
    
    for yr in years:
        # '사업체수 (개)' 컬럼의 데이터 추출
        val = incheon_row[(str(yr), '사업체수 (개)', '계')]
        biz_counts.append(float(val))

    # 3. 시각화 (이중 축 그래프)
    fig, ax1 = plt.subplots(figsize=(12, 7), constrained_layout=True)

    # 축 1: 송도 개최 건수 (막대 그래프)
    bar = ax1.bar(years, df_s_filtered['개최건수'], color='skyblue', alpha=0.5, label='송도 전시 개최건수')
    ax1.set_xlabel('연도', fontsize=12)
    ax1.set_ylabel('개최 건수 (건)', color='dodgerblue', fontsize=12, fontweight='bold')
    ax1.tick_params(axis='y', labelcolor='dodgerblue')

    # 축 2: 인천 사업체 수 (꺾은선 그래프)
    ax2 = ax1.twinx()
    line = ax2.plot(years, biz_counts, marker='s', color='red', linewidth=3, markersize=8, label='인천 전시 사업체수')
    ax2.set_ylabel('사업체 수 (개)', color='red', fontsize=12, fontweight='bold')
    ax2.tick_params(axis='y', labelcolor='red')

    # 데이터 라벨 표시 (개최건수)
    for i, val in enumerate(df_s_filtered['개최건수']):
        ax1.text(years[i], val + 1, f'{int(val)}건', ha='center', va='bottom', color='dodgerblue', fontweight='bold')

    # 데이터 라벨 표시 (사업체수)
    for i, val in enumerate(biz_counts):
        ax2.text(years[i], val + (max(biz_counts)*0.02), f'{int(val)}개', ha='center', va='bottom', color='red', fontweight='bold')

    # 제목 및 그리드
    plt.title('인천 전시 사업체 수와 송도 전시 개최건수 상관관계 (2018-2023)', fontsize=16, fontweight='bold', pad=20)
    ax1.grid(True, axis='y', linestyle='--', alpha=0.3)

    # 범례 통합 표시
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines + lines2, labels + labels2, loc='upper left')

    # y축 범위 최적화 (라벨 안잘리게)
    ax1.set_ylim(0, max(df_s_filtered['개최건수']) * 1.2)
    ax2.set_ylim(min(biz_counts) * 0.8, max(biz_counts) * 1.2)

    plt.show()

except Exception as e:
    print(f"오류 발생: {e}")
    print("파일 이름과 시트 구조를 다시 한번 확인해주세요.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import platform

# 1. 한글 폰트 및 마이너스 기호 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
else:
    plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 로드
file_songdo = '송도 연별 개최건수.xlsx'
file_employment = '전시사업자_업종_및_지역별__고용형태별_2018_2023.xlsx'

try:
    # --- 데이터 1: 송도 개최 건수 정리 ---
    df_s = pd.read_excel(file_songdo)
    df_s_trend = df_s.set_index(df_s.columns[0]).T.reset_index()
    df_s_trend.columns = ['연도', '개최건수']
    df_s_trend['연도'] = df_s_trend['연도'].astype(int)
    # 2018~2023년 필터링
    df_s_filtered = df_s_trend[(df_s_trend['연도'] >= 2018) & (df_s_trend['연도'] <= 2023)]

    # --- 데이터 2: 인천 정규직/비정규직 종사자 데이터 정리 ---
    # 이 파일은 컬럼 구조가 매우 복잡하므로 위치(iloc) 기반으로 인천 데이터를 추출합니다.
    df_e = pd.read_excel(file_employment)
    
    # '인천' 행 찾기 (두 번째 컬럼 '특성별(2)' 기준)
    incheon_data = df_e[df_e.iloc[:, 1].str.contains('인천', na=False)].iloc[0]
    
    years = [2018, 2019, 2020, 2021, 2022, 2023]
    regular_list = []      # 정규직
    non_regular_list = []  # 비정규직 (임시 및 일용직 등)

    # 엑셀 구조상 연도별로 12개의 컬럼이 반복됨 (계, 정규직, 비정규직 등)
    # 2018년 시작 인덱스는 2번부터이며, 각 연도별 '정규직'과 '비정규직' 위치를 지정
    start_col = 2
    for i in range(len(years)):
        idx = start_col + (i * 12)
        # 엑셀 헤더 순서에 따른 인덱스: 정규직(idx+1), 비정규직(idx+4: 임시 및 일용직 합계 등)
        # 데이터의 정확한 위치는 파일의 '계' 하위 항목 순서를 따름
        reg = float(incheon_data.iloc[idx + 1])      # 정규직 계
        non_reg = float(incheon_data.iloc[idx + 4])  # 비정규직(임시 및 일용직) 계
        
        regular_list.append(reg)
        non_regular_list.append(non_reg)

    # 3. 시각화 (이중 축 그래프)
    fig, ax1 = plt.subplots(figsize=(12, 7), constrained_layout=True)

    # 축 1: 고용 형태별 종사자 수 (누적 막대 그래프)
    p1 = ax1.bar(years, regular_list, color='#3498db', label='정규직 종사자', alpha=0.8)
    p2 = ax1.bar(years, non_regular_list, bottom=regular_list, color='#e74c3c', label='비정규직 종사자', alpha=0.8)
    
    ax1.set_ylabel('종사자 수 (명)', fontsize=12, fontweight='bold')
    ax1.set_xlabel('연도', fontsize=12)
    ax1.grid(axis='y', linestyle='--', alpha=0.5)

    # 축 2: 전시회 개최 건수 (꺾은선 그래프)
    ax2 = ax1.twinx()
    line = ax2.plot(years, df_s_filtered['개최건수'], marker='o', color='black', 
                    linewidth=3, markersize=10, label='전시회 개최건수')
    ax2.set_ylabel('개최 건수 (건)', fontsize=12, fontweight='bold', rotation=270, labelpad=20)

    # 수치 표시 (개최건수 강조)
    for i, val in enumerate(df_s_filtered['개최건수']):
        ax2.text(years[i], val + 2, f'{int(val)}건', ha='center', va='bottom', 
                 fontsize=11, fontweight='bold', bbox=dict(facecolor='white', alpha=0.5, edgecolor='none'))

    # 제목 및 범례
    plt.title('인천 전시사업자 고용형태와 송도 전시 개최건수 추이 (2018-2023)', fontsize=16, fontweight='bold', pad=20)
    
    # 범례 합치기
    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax2.get_legend_handles_labels()
    ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

    plt.show()

except Exception as e:
    print(f"오류 발생: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import platform

# 1. 한글 폰트 설정 (환경에 맞게 자동 선택)
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
else:
    plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 경로 설정 (본인의 실제 파일 경로로 수정하세요)
# 여기서는 요청하신 대로 .xlsx 확장자를 사용합니다.
file_songdo = '송도 연별 개최건수.xlsx'
file_employment = '전시사업자_업종_및_지역별__고용형태별_2018_2023.xlsx'

try:
    # --- 데이터 1: 송도 개최 건수 (Excel 읽기) ---
    # 시트 이름이 다를 경우 sheet_name을 수정하세요.
    df_s_raw = pd.read_excel(file_songdo) 
    
    # 데이터 구조 변환 (연도가 컬럼으로 되어 있는 경우 행으로 변경)
    df_s_trend = df_s_raw.set_index(df_s_raw.columns[0]).T.reset_index()
    df_s_trend.columns = ['연도', '개최건수']
    df_s_trend['연도'] = df_s_trend['연도'].astype(int)
    
    # 2018~2023년 데이터만 추출
    df_s_filtered = df_s_trend[(df_s_trend['연도'] >= 2018) & (df_s_trend['연도'] <= 2023)]

    # --- 데이터 2: 인천 고용 데이터 (Excel 읽기) ---
    df_e = pd.read_excel(file_employment)
    
    # '인천' 지역 데이터 추출 (두 번째 열에 '인천'이 포함된 행)
    incheon_row = df_e[df_e.iloc[:, 1].str.contains('인천', na=False)].iloc[0]
    
    years = [2018, 2019, 2020, 2021, 2022, 2023]
    total_employment = []

    # 엑셀 열 구조에 맞춰 고용 합계(정규직+비정규직) 계산
    # 보통 통계청 데이터는 연도별로 일정한 간격(예: 12칸)을 두고 데이터가 반복됩니다.
    start_col = 2
    for i in range(len(years)):
        idx = start_col + (i * 12)
        reg = float(incheon_row.iloc[idx + 1])      # 정규직 계
        non_reg = float(incheon_row.iloc[idx + 4])  # 비정규직 계
        total_employment.append(reg + non_reg)

    # 3. 시각화 시작
    fig, ax1 = plt.subplots(figsize=(12, 7))

    # [배경] 축 1: 전시회 개최 건수 (연한 회색 막대그래프)
    ax1.bar(years, df_s_filtered['개최건수'], color='#D3D3D3', alpha=0.5, label='전시회 개최건수', zorder=1)
    ax1.set_ylabel('전시회 개최 건수 (건)', fontsize=12, color='#7F7F7F')
    ax1.tick_params(axis='y', labelcolor='#7F7F7F')
    ax1.set_xlabel('연도', fontsize=12)

    # [강조] 축 2: 총 종사자 수 (진한 남색 꺾은선그래프)
    ax2 = ax1.twinx()
    ax2.plot(years, total_employment, marker='s', color='#1F4E78', 
             linewidth=3, markersize=10, label='인천 전시 종사자 수(총계)', zorder=2)
    ax2.set_ylabel('총 종사자 수 (명)', fontsize=12, fontweight='bold')

    # 수치 표시 (꺾은선 그래프 위)
    for i, val in enumerate(total_employment):
        ax2.text(years[i], val + 15, f'{int(val)}명', ha='center', va='bottom', 
                 fontsize=11, fontweight='bold', color='#1F4E78')

    # 제목 및 디자인 정리
    plt.title('송도 전시회 개최건수 대비 인천 고용 인원 추이 (2018-2023)', fontsize=16, pad=20)
    
    # 범례 통합
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

    ax1.grid(axis='y', linestyle=':', alpha=0.5)
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"오류가 발생했습니다: {e}")
    print("팁: openpyxl 라이브러리가 설치되어 있는지 확인하세요 (pip install openpyxl)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import platform

# 1. 한글 폰트 및 마이너스 기호 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
else:
    plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 파일 로드
file_songdo = '송도 연별 개최건수.xlsx'
file_employment = '전시사업자_업종_및_지역별__고용형태별_2018_2023.xlsx'

try:
    # --- 데이터 1: 송도 개최 건수 정리 ---
    df_s_raw = pd.read_excel(file_songdo)
    df_s_trend = df_s_raw.set_index(df_s_raw.columns[0]).T.reset_index()
    df_s_trend.columns = ['연도', '개최건수']
    df_s_trend['연도'] = df_s_trend['연도'].astype(int)
    df_s_filtered = df_s_trend[(df_s_trend['연도'] >= 2018) & (df_s_trend['연도'] <= 2023)]

    # --- 데이터 2: 인천 정규직/비정규직/총계 데이터 정리 ---
    df_e = pd.read_excel(file_employment)
    incheon_row = df_e[df_e.iloc[:, 1].str.contains('인천', na=False)].iloc[0]
    
    years = [2018, 2019, 2020, 2021, 2022, 2023]
    regular_list = []      # 정규직
    non_regular_list = []  # 비정규직
    total_list = []        # 총계

    start_col = 2
    for i in range(len(years)):
        idx = start_col + (i * 12)
        reg = float(incheon_row.iloc[idx + 1])      # 정규직
        non_reg = float(incheon_row.iloc[idx + 4])  # 비정규직
        
        regular_list.append(reg)
        non_regular_list.append(non_reg)
        total_list.append(reg + non_reg)

    # 3. 시각화 (이중 축 그래프)
    # constrained_layout=True를 사용하여 여백을 자동 조절합니다.
    fig, ax1 = plt.subplots(figsize=(14, 8), constrained_layout=True)

    # [배경] 축 1: 전시회 개최 건수 (연한 회색 막대)
    ax1.bar(years, df_s_filtered['개최건수'], color='#E0E0E0', alpha=0.5, label='전시회 개최건수', zorder=1)
    ax1.set_ylabel('전시회 개최 건수 (건)', fontsize=12, color='#999999', fontweight='bold')
    ax1.tick_params(axis='y', labelcolor='#999999')
    ax1.set_xlabel('연도', fontsize=12)

    # [강조] 축 2: 고용 형태별 추이 (총계, 정규직, 비정규직)
    ax2 = ax1.twinx()
    
    # 1. 총계: 검은색 실선
    line_total = ax2.plot(years, total_list, marker='D', color='#2D3436', 
                          linewidth=3, markersize=7, label='총 종사자 수', zorder=5)
    # 2. 정규직: 파란색 실선
    line_reg = ax2.plot(years, regular_list, marker='o', color='#1F77B4', 
                        linewidth=2, markersize=7, label='정규직 종사자 수', zorder=4)
    # 3. 비정규직: 오렌지색 대시선
    line_non = ax2.plot(years, non_regular_list, marker='s', color='#FF7F0E', 
                        linewidth=2, markersize=7, linestyle='--', label='비정규직 종사자 수', zorder=4)
    
    ax2.set_ylabel('종사자 수 (명)', fontsize=12, fontweight='bold')

    # 수치 표시 (겹침 방지를 위해 오프셋 미세 조절)
    for i in range(len(years)):
        # 총계 (위쪽)
        ax2.text(years[i], total_list[i] + 15, f'{int(total_list[i])}', ha='center', va='bottom', color='#2D3436', fontweight='bold')
        # 정규직 (값에 따라 위치 조정)
        ax2.text(years[i], regular_list[i] + 10, f'{int(regular_list[i])}', ha='center', va='bottom', color='#1F77B4', fontsize=9)
        # 비정규직 (아래쪽)
        ax2.text(years[i], non_regular_list[i] - 20, f'{int(non_regular_list[i])}', ha='center', va='top', color='#FF7F0E', fontsize=9)

    # 제목 설정
    plt.title('송도 전시회 개최건수 및 고용 형태별(정규/비정규/총계) 추이 (2018-2023)', fontsize=16, pad=30, fontweight='bold')
    
    # [핵심 수정] 범례와 그래프 겹침 방지
    # 1. 범례를 그래프 밖 상단으로 빼거나, Y축 범위를 넓혀 공간 확보
    ax1.set_ylim(0, max(df_s_filtered['개최건수']) * 1.6) # 막대 그래프 상단 여백 확보
    ax2.set_ylim(min(non_regular_list) * 0.7, max(total_list) * 1.3) # 선 그래프 상단 여백 확보

    # 범례 통합 표시 (loc='upper left' 유효, 여백 확보로 겹치지 않음)
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', frameon=True, shadow=True, ncol=2)

    ax1.grid(axis='y', linestyle=':', alpha=0.5)
    plt.show()

except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings

# openpyxl 관련 스타일 경고 무시
warnings.filterwarnings('ignore', category=UserWarning, module='openpyxl')

# 1. 파일 이름 설정
file_sales = '전시사업자_업종_및_지역별_전시_관련_매출액__영업이익_20260410110142.xlsx'
file_count = '송도 연별 개최건수.xlsx'

try:
    # 2. 데이터 불러오기
    df_sales = pd.read_excel(file_sales, header=[0, 1])
    df_count = pd.read_excel(file_count)

    # 3. 영업이익 데이터 전처리 (인천 지역 추출)
    incheon_mask = df_sales.iloc[:, 1].astype(str).str.contains('인천')
    incheon_row = df_sales[incheon_mask].iloc[0]
    
    years_str = ['2018', '2019', '2020', '2021', '2022', '2023']
    profit_values = [] # 영업이익 저장용
    
    for yr in years_str:
        # '영업이익 (백만원)' 컬럼 값 추출
        try:
            val = incheon_row[(yr, '영업이익 (백만원)')]
        except KeyError:
            val = incheon_row[(int(yr), '영업이익 (백만원)')]
        
        # 숫자 변환 (결측치 및 문자열 예외 처리)
        try:
            profit_values.append(float(val))
        except (ValueError, TypeError):
            profit_values.append(0)

    # 4. 송도 개최 건수 데이터 전처리
    df_count_indexed = df_count.set_index(df_count.columns[0])
    df_count_indexed.columns = df_count_indexed.columns.astype(str)
    
    count_values = []
    for yr in years_str:
        if yr in df_count_indexed.columns:
            val_count = df_count_indexed.loc['전시회 개최수', yr]
            count_values.append(int(val_count) if pd.notna(val_count) else 0)
        else:
            count_values.append(0)

    # 5. 이중 축 시각화
    plt.rcParams['font.family'] = 'Malgun Gothic'
    plt.rcParams['axes.unicode_minus'] = False

    fig, ax1 = plt.subplots(figsize=(12, 6))

    # --- 왼쪽 축: 인천 영업이익 (Green) ---
    color_profit = '#2ca02c' # 녹색
    ax1.set_xlabel('연도')
    ax1.set_ylabel('인천 전시 영업이익 (백만 원)', color=color_profit, fontweight='bold')
    lns1 = ax1.plot(years_str, profit_values, marker='o', color=color_profit, label='인천 전시 영업이익', linewidth=2.5)
    ax1.tick_params(axis='y', labelcolor=color_profit)
    
    # 영업이익 0선(기준선) 추가
    ax1.axhline(0, color='gray', linestyle='-', linewidth=1, alpha=0.5)

    # --- 오른쪽 축: 개최 건수 (Red) ---
    color_count = '#d62728' # 빨간색
    ax2 = ax1.twinx()
    ax2.set_ylabel('송도 전시회 개최 건수 (건)', color=color_count, fontweight='bold')
    lns2 = ax2.plot(years_str, count_values, marker='s', color=color_count, label='송도 개최 건수', linestyle='--', linewidth=2)
    ax2.tick_params(axis='y', labelcolor=color_count)

    # 오른쪽 y축 범위 및 눈금 최적화
    max_count = max(count_values)
    ax2.set_ylim(0, max_count + 15)
    ax2.set_yticks(np.arange(0, max_count + 20, 10))

    # 데이터 라벨 표시 (개최 건수 강조)
    for i, txt in enumerate(count_values):
        ax2.annotate(f'{txt}', (years_str[i], count_values[i]), 
                     textcoords="offset points", xytext=(0,10), ha='center', color=color_count, fontweight='bold')

    # 범례 및 제목
    plt.title('인천 전시 영업이익 및 송도 개최 건수 상관 분석 (2018-2023)', fontsize=14, pad=20)
    lns = lns1 + lns2
    labs = [l.get_label() for l in lns]
    ax1.legend(lns, labs, loc='upper left')

    plt.grid(True, axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"오류 발생: {e}")
```

```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time

# 1. 데이터 불러오기
df = pd.read_csv('전국_철강산업_공장현황.csv', encoding='utf-8-sig')

# 테스트를 위해 상위 100개 데이터만 사용 (전체는 시간이 오래 걸릴 수 있음)
df_sample = df.head(100).copy()

# 2. 주소를 좌표로 변환 (Geocoding)
geolocator = Nominatim(user_agent="south_korea_steel_map")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

print("주소를 좌표로 변환 중입니다. 잠시만 기다려 주세요...")

# '공장주소' 컬럼을 기반으로 좌표 추출
df_sample['location'] = df_sample['공장주소'].apply(geocode)
df_sample['lat'] = df_sample['location'].apply(lambda loc: loc.latitude if loc else None)
df_sample['lon'] = df_sample['location'].apply(lambda loc: loc.longitude if loc else None)

# 좌표 변환에 실패한 행 제거
df_sample = df_sample.dropna(subset=['lat', 'lon'])

# 3. 대한민국 지도 생성
# 지도의 중심을 대한민국 중앙으로 설정
m = folium.Map(location=[36.5, 127.5], zoom_start=7, tiles='cartodbpositron')

# 4. 버블 마커 추가
for i, row in df_sample.iterrows():
    # 종업원 수에 따라 버블 크기 결정 (데이터 내 '종업원합계' 컬럼 활용)
    # 값이 클 경우를 대비해 루트나 로그를 씌우면 더 보기 좋습니다.
    pop_size = row['종업원합계'] if pd.notnull(row['종업원합계']) else 5
    
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=pop_size / 5,  # 크기 조절 계수
        color='#e74c3c',      # 테두리 색상 (강렬한 철강의 느낌 - 빨간색)
        fill=True,
        fill_color='#c0392b', # 채우기 색상
        fill_opacity=0.6,
        popup=folium.Popup(f"<b>{row['회사명']}</b><br>생산품: {row['생산품']}<br>종업원: {row['종업원합계']}명", max_width=300)
    ).add_to(m)

# 5. 결과 저장 및 확인
m.save('대한민국_철강산업_분포_버블맵.html')
print("버블맵 생성이 완료되었습니다. '대한민국_철강산업_분포_버블맵.html' 파일을 확인하세요.")
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import requests
import time

# 1. 네이버 API 키 설정 (본인의 키로 수정하세요)
CLIENT_ID = "tRxyx2FZijBtjwOIGPgG"
CLIENT_SECRET = "7um7OvF2bL"


# 2. 주소를 좌표로 변환하는 함수 (네이버 Geocoding API)
def get_coords_naver(address):
    url = f"https://openapi.naver.com/v1/map-geocode/search?query={address}"
    headers = {
        "X-Naver-Client-Id": CLIENT_ID,
        "X-Naver-Client-Secret": CLIENT_SECRET
    }
    try:
        response = requests.get(url, headers=headers).json()
        if 'addresses' in response and len(response['addresses']) > 0:
            # 위도(y), 경도(x) 반환
            return float(response['addresses'][0]['y']), float(response['addresses'][0]['x'])
    except Exception as e:
        return None, None
    return None, None


# 3. 데이터 불러오기
file_path = '전국_철강산업_공장현황.csv'
df = pd.read_csv(file_path, encoding='utf-8-sig')

# 4. 전체 데이터 좌표 변환
# 주의: 데이터가 많으면 수십 분이 소요될 수 있습니다.
# 테스트를 원하시면 df = df.head(100)을 먼저 실행해 보세요.
print(f"총 {len(df)}건의 데이터를 변환하기 시작합니다...")

lats, lons = [], []
for i, addr in enumerate(df['공장주소']):
    lat, lon = get_coords_naver(addr)
    lats.append(lat)
    lons.append(lon)

    # 진행 상황 출력 (100건마다)
    if (i + 1) % 100 == 0:
        print(f"진행 중... ({i + 1}/{len(df)})")

df['lat'] = lats
df['lon'] = lons

# 좌표 변환에 성공한 데이터만 추출
df_final = df.dropna(subset=['lat', 'lon'])

# 5. 버블맵 시각화
print("지도를 생성하고 있습니다...")
m = folium.Map(location=[36.5, 127.5], zoom_start=7, tiles='cartodbpositron')

# 마커 클러스터 추가 (데이터가 많으므로 필수)
marker_cluster = MarkerCluster().add_to(m)

for i, row in df_final.iterrows():
    # 종업원 수 기반 버블 크기 설정 (값이 너무 크면 루트 처리)
    emp_count = row['종업원합계'] if pd.notnull(row['종업원합계']) else 5
    radius = (emp_count ** 0.5) * 1.5  # 크기 조정 계수

    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=radius if radius > 2 else 2,
        color='#e74c3c',  # 테두리: 빨강
        fill=True,
        fill_color='#c0392b',  # 채우기: 진한 빨강
        fill_opacity=0.5,
        popup=folium.Popup(f"<b>{row['회사명']}</b><br>생산품: {row['생산품']}<br>종업원: {row['종업원합계']}명", max_width=300)
    ).add_to(marker_cluster)

# 6. 결과 저장
m.save('전국_철강산업_전체_버블맵.html')
print(f"성공적으로 {len(df_final)}개의 공장을 지도에 표시했습니다.")
print("'전국_철강산업_전체_버블맵.html' 파일을 웹 브라우저로 열어보세요.")
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. 데이터 불러오기
df = pd.read_csv('전국_철강산업_공장현황.csv')

# 2. 행정구역별 데이터 집계 (시도별 공장 수 및 종업원 수)
agg_df = df.groupby('시도명').agg({
    '공장관리번호': 'count',
    '종업원합계': 'sum'
}).reset_index()
agg_df.columns = ['시도명', '공장수', '종업원합계']

# 3. 대한민국 행정구역 경계 데이터(GeoJSON) 불러오기
# 인터넷에서 한국 행정구역 GeoJSON 파일을 다운로드하여 사용합니다.
url = "https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2013/json/skorea_provinces_geo.json"
korea_map = gpd.read_file(url)

# 4. 지도 데이터와 통계 데이터 결합 (이름 매칭)
# GeoJSON의 'name' 컬럼과 CSV의 '시도명'을 일치시킵니다.
korea_map = korea_map.merge(agg_df, left_on='name', right_on='시도명', how='left')

# 5. 각 시도의 중심점(Centroid) 계산 (버블을 찍을 위치)
korea_map['centroid'] = korea_map.geometry.centroid

# 6. 시각화
fig, ax = plt.subplots(1, 1, figsize=(10, 12))

# (1) 대한민국 기본 지도 그리기
korea_map.plot(ax=ax, color='whitesmoke', edgecolor='lightgrey', linewidth=1)

# (2) 버블맵 그리기
# 버블 크기(s): 공장 수, 버블 색상(c): 종업원 수
scatter = ax.scatter(
    korea_map.centroid.x, 
    korea_map.centroid.y, 
    s=korea_map['공장수'] * 1.2, # 공장 수에 따른 크기 조절
    c=korea_map['종업원합계'],   # 종업원 수에 따른 색상 조절
    cmap='OrRd',               # 오렌지-빨강 컬러맵
    alpha=0.7, 
    edgecolors='darkgrey', 
    linewidth=0.5
)

# (3) 지역명 표시
for idx, row in korea_map.iterrows():
    ax.annotate(text=row['name'], xy=(row.centroid.x, row.centroid.y),
                xytext=(0, 5), textcoords="offset points",
                ha='center', fontsize=9, color='black')

# 7. 차트 꾸미기
plt.colorbar(scatter, label='총 종업원 수 (Color)')
plt.title('전국 철강산업 공장 분포 및 고용 현황', fontsize=16, pad=20)
ax.set_axis_off() # 위도/경도 축 숨기기

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. 데이터 불러오기
df = pd.read_csv('전국_철강산업_공장현황.csv')

# 2. 시군구별 데이터 집계 및 컬럼명 변경 (이 부분이 수정되었습니다)
agg_df = df.groupby(['시도명', '시군구명']).agg({
    '공장관리번호': 'count',
    '종업원합계': 'sum'
}).reset_index()

# 컬럼명을 '공장수'와 '종업원합계'로 명확하게 지정합니다.
agg_df.columns = ['시도명', '시군구명', '공장수', '종업원합계']

# 3. 대한민국 시군구 행정구역 경계 데이터(GeoJSON) 불러오기
url = "https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2013/json/skorea_municipalities_geo.json"
korea_map = gpd.read_file(url)

# 4. 지도 데이터와 통계 데이터 결합
# GeoJSON의 'name'(시군구명)과 CSV의 '시군구명'을 매칭
korea_map = korea_map.merge(agg_df, left_on='name', right_on='시군구명', how='left')

# 결측치(공장이 없는 지역) 처리
korea_map['공장수'] = korea_map['공장수'].fillna(0)
korea_map['종업원합계'] = korea_map['종업원합계'].fillna(0)

# 5. 각 시군구의 중심점(Centroid) 계산
korea_map['centroid'] = korea_map.geometry.centroid

# 6. 시각화
fig, ax = plt.subplots(1, 1, figsize=(12, 15))

# 배경 지도
korea_map.plot(ax=ax, color='whitesmoke', edgecolor='silver', linewidth=0.5)

# 버블맵
scatter = ax.scatter(
    korea_map.centroid.x, 
    korea_map.centroid.y, 
    s=korea_map['공장수'] * 10,   # 시군구 단위는 버블이 작으므로 배율을 높임 (10)
    c=korea_map['종업원합계'], 
    cmap='YlOrRd', 
    alpha=0.6, 
    edgecolors='grey', 
    linewidth=0.3
)

# 상위 15개 지역 라벨링 (포항, 광양 등 주요 지역 강조)
top_areas = korea_map.nlargest(15, '공장수')
for idx, row in top_areas.iterrows():
    ax.annotate(text=row['시군구명'], xy=(row.centroid.x, row.centroid.y),
                xytext=(0, 8), textcoords="offset points",
                ha='center', fontsize=9, color='darkred', weight='bold')

# 차트 정리
cbar = plt.colorbar(scatter, ax=ax, shrink=0.5)
cbar.set_label('시군구별 총 종업원 수')
plt.title('전국 시군구별 철강산업 분포 현황', fontsize=18, pad=20)
ax.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import folium

# 1. 데이터 불러오기
file_path = '전국_철강산업_공장현황.csv'
df = pd.read_csv(file_path, encoding='utf-8-sig')

# 2. [필터링] 대기업만 추출
df_large = df[df['공장규모'] == '대기업'].copy()

# 3. 포항시 남구/북구를 '포항시'로 통합
def clean_city_name(row):
    city = str(row['시군구명'])
    if '포항시' in city:
        return '포항시'
    return city

df_large['통합시군구'] = df_large.apply(clean_city_name, axis=1)
df_large['지역'] = df_large['시도명'] + " " + df_large['통합시군구']

# 4. 지역별 대기업 공장 수 집계
city_counts = df_large['지역'].value_counts().reset_index()
city_counts.columns = ['지역', '공장수']

# 5. 지도 생성
m = folium.Map(location=[36.5, 127.5], zoom_start=7, tiles='CartoDB positron')

# 6. 주요 대기업 공장 소재지 좌표 (시군구 단위)
geo_cities = {
    '경상북도 포항시': [36.019, 129.343],
    '전라남도 광양시': [34.940, 127.690],
    '충청남도 당진시': [36.892, 126.629],
    '인천광역시 서구': [37.545, 126.675],
    '울산광역시 남구': [35.544, 129.331],
    '부산광역시 사하구': [35.084, 128.981],
    '전라남도 순천시': [34.950, 127.487],
    '경상남도 창원시': [35.227, 128.681],
    '충청남도 천안시': [36.815, 127.113],
    '경기도 평택시': [36.992, 127.112]
}

# 7. 버블 추가
for index, row in city_counts.iterrows():
    area = row['지역']
    count = row['공장수']
    
    if area in geo_cities:
        # 대기업은 절대적인 수가 적으므로 반지름(radius)을 
        # '공장수 * 5' 정도로 키워 시인성을 확보합니다.
        folium.CircleMarker(
            location=geo_cities[area],
            radius=count * 5,  
            popup=f"<b>{area} (대기업)</b><br>공장수: {count}개",
            color='#2c3e50',      # 대기업의 무게감을 주는 남색 테두리
            fill=True,
            fill_color='#2980b9', # 신뢰감을 주는 파란색 채우기
            fill_opacity=0.7,
            weight=3
        ).add_to(m)

# 8. 결과 저장
m.save('전국_철강_대기업_분포맵.html')
print(f"대기업 필터링 완료! 총 {len(df_large)}개의 대기업 공장이 집계되었습니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 불러오기
file_path = '전국_철강산업_공장현황.csv'
df = pd.read_csv(file_path, encoding='utf-8-sig')

# 2. 지역별(시도명) 공장 수 집계 및 상위 5개 추출 (전체 데이터 사용)
top5_regions = df['시도명'].value_counts().head(5)

# 3. 시각화 설정 (한글 깨짐 방지)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 4. 가로 막대 그래프 그리기 (상위 지역이 위로 오도록 정렬)
top5_sorted = top5_regions.sort_values(ascending=True)

plt.figure(figsize=(10, 6))

# 두 번째 코드에서 사용된 붉은색 계열 컬러 팔레트 적용
colors = ['#ef9a9a', '#e57373', '#f44336', '#ef5350', '#d32f2f']
bars = plt.barh(top5_sorted.index, top5_sorted.values, color=colors, edgecolor='black', alpha=0.9)

# 그래프 오른쪽에 수치 표시
for bar in bars:
    width = bar.get_width()
    plt.text(width + (top5_regions.max() * 0.01), bar.get_y() + bar.get_height()/2, 
             f'{int(width)}개', va='center', fontweight='bold', fontsize=11)

# 타이틀 및 라벨 설정
plt.title('전국 철강 산업 공장 수 상위 5개 지역 (전체)', fontsize=16, pad=20, fontweight='bold')
plt.xlabel('공장 수 (개)', fontsize=12)
plt.ylabel('지역(시도)', fontsize=12)

# x축 범위에 여유 주기 (텍스트 공간 확보)
plt.xlim(0, top5_regions.max() * 1.1)

# 그리드 추가 (x축 기준)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
