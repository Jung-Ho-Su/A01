```python
import pandas as pd

# read_sav를 read_spss로 변경
df = pd.read_spss('HN24_ALL.sav')

print(df.head())
```

```python
df.keys()
```

```python
print(df.columns.tolist())
for col in df.columns:
    print(col)
```

```python
# 'korean_data.csv'라는 이름으로 저장 (인덱스 제외, 한글 깨짐 방지 설정)
df.to_csv('HN24_ALL.csv', index=False, encoding='utf-8-sig')
```
