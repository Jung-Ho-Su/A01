```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 1. 한글 폰트 설정 (윈도우 기준 'Malgun Gothic' 사용)
# 만약 맥을 사용하신다면 'AppleGothic'으로 수정하세요.
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 2. 데이터 불러오기
file_path = "서울교통공사_23년도_시간대별_전체총합.csv"
df = pd.read_csv(file_path, encoding='utf-8-sig')

# 3. 시각화용 데이터 가공
# '구분' 컬럼을 제외한 시간대 컬럼들만 리스트로 뽑습니다.
time_cols = [col for col in df.columns if '시' in col]
# 데이터프레임에서 숫자 값들만 추출하여 리스트로 만듭니다.
passenger_values = df[time_cols].iloc[0].values

# 4. 꺾은선 그래프 그리기
plt.figure(figsize=(15, 7)) # 그래프 크기 설정
plt.plot(time_cols, passenger_values, marker='o', linestyle='-', color='#1f77b4', linewidth=2)

# 5. 그래프 꾸미기 (제목, 축 레이블 등)
plt.title('2023년 서울 지하철 시간대별 전체 이용객 합계', fontsize=16, pad=20)
plt.xlabel('시간대', fontsize=12)
plt.ylabel('총 이용객 수 (억 단위)', fontsize=12)

# X축 글자 기울이기 (시간대 이름이 겹치지 않게 45도 회전)
plt.xticks(rotation=45)

# 그리드(격자) 추가
plt.grid(True, linestyle='--', alpha=0.6)

# 레이아웃 자동 조정
plt.tight_layout()

# 6. 이미지로 저장 및 화면 표시
plt.savefig('subway_congestion_trend.png', dpi=300)
plt.show()

print("✅ 그래프 생성이 완료되었습니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 1. 한글 폰트 설정 (윈도우 환경에서 가장 안전한 방법)
def set_korean_font():
    # 윈도우 기본 폰트인 '맑은 고딕'을 우선 설정합니다.
    font_list = [f.name for f in fm.fontManager.ttflist]
    if 'Malgun Gothic' in font_list:
        plt.rcParams['font.family'] = 'Malgun Gothic'
    elif 'NanumGothic' in font_list:
        plt.rcParams['font.family'] = 'NanumGothic'
    else:
        # 설치된 폰트 중 'Gothic'이 포함된 폰트를 아무거나 찾아 설정 (최후의 수단)
        for font in font_list:
            if 'Gothic' in font:
                plt.rcParams['font.family'] = font
                break
    
    plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

set_korean_font()

# 2. 데이터 불러오기
file_path = "서울교통공사_24년도_승차시간별_전체총합.csv"

try:
    df = pd.read_csv(file_path, encoding='cp949')
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding='utf-8-sig')

# 3. 시각화용 데이터 가공
# '구분' 컬럼 제외하고 시간대 컬럼만 추출
time_cols = [col for col in df.columns if '시' in col]
boarding_values = df[time_cols].iloc[0].values

# 4. 꺾은선 그래프 그리기
plt.figure(figsize=(15, 7))
plt.plot(time_cols, boarding_values, marker='o', linestyle='-', color='#1f77b4', linewidth=2, label='2024년 승차 총합')

# 5. 그래프 꾸미기
plt.title('2024년 서울 지하철 시간대별 전체 승차 인원 추이', fontsize=16, pad=20)
plt.xlabel('시간대', fontsize=12)
plt.ylabel('승차 인원 수', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# 6. 이미지 저장 및 출력
plt.tight_layout()
plt.savefig('subway_2024_boarding_trend.png', dpi=300)
plt.show()

print("✅ 그래프 생성이 완료되었습니다.")
```
