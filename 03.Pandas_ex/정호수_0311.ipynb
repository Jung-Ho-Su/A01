```python
# 문제 1 / 아래 출력 결과를 참고하여 DataFrame을 생성하여 df라는 이름으로 저장하는 코드를 작성하시오.
import pandas as pd

data = {
    "food" : ["KFC","McDonald","SchoolFood"],
    "price" : [1000, 2000, 2500],
    "rating" : [4.5, 3.9, 4.2]
}

column_list = ["food", "price", "rating"]
index_list = [0, 1, 2]
df = pd.DataFrame(data, index=index_list, columns=column_list)
df
```

```python
# 문제 2 / 문제 1에서 foor 컬럼과 rating 컬럼만 선택하여 출력하는 코드를 작성하시오.
df[['food','rating']]
```

```python
# 문제 3 / 문제 1에서 food 컬럼명을 place로 컬럼명을 변경하는 코드를 작성하시오.
df2 = df.rename(columns={"food":"place"})
df2
```
