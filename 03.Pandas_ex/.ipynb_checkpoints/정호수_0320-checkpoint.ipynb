# [문제] Seaborn이 제공하고 있는 데이터 셋을 사용하여 질문에 답하시오.

- (1) 팁 데이터에서 pivot_table()함수를 사용하여 index='day', columns='size', values='tip'으로하는 데이터를 만든 다음 해당 데이터로 히트맵을 작성하시오.


- (2) 아이리스 데이터에서 각 데이터의 상관관계를 pairplot()으로 나타내시오.

```python
import ssl
# 전역적으로 SSL 검증을 하지 않도록 설정
ssl._create_default_https_context = ssl._create_unverified_context
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')

tips_size = tips.pivot_table(index='day', columns='size', values='tip')
tips_size
```

```python
sns.heatmap(tips_size, 
            cmap=sns.light_palette('gray', as_cmap=True), 
            annot=True)
plt.title('Heatmap')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
plt.rcParams['font.family'] = 'Malgun Gothic'
sns.pairplot(tips)
plt.title('tips Data의 Pair Plot')
plt.show()
```
