```python
from pandas import Series

# data를 series의 형식으로 인덱스, 값 의 콜롬(열)을 가지도록 만들어주고 
# 인덱스에 관한 정보를 출력하도록 함

data = [1000, 2000, 3000]
s = Series(data)
print(s.index)
print(s.index.to_list())  # to_list() 로 출력하게 되면 마지막에 list의 형식으로 추력할 수 있다
```

```python
data = [1000, 2000, 3000]
s = Series(data) # 각 데이터 내용을 정수형태로 변환
s.index = ["메로나", "구구콘", "하겐다즈"]
print(s)
```

```python
data = [1000, 2000, 3000]
index = ["메로나", "구구콘", "하겐다즈"]  # 위 cell과 같음 하지만 keyword 형식으로 주어진 것 
s = Series(data=data, index=index)  
print(s)                     

# s = Series(data, index)로 생략해서 표현 가능 / s = Series(data, index=index) 앞에는 일반 파라매터 뒤는 keyword 파라매터로 잡아줌
# s = Series(data=data, index) 얘는 실행 불가함! ㄴ왜냐면 앞에 keyword 매개변수가 먼저 와서 안댐 일반 매개변수가 먼저 와야함

# series classs에서 데이터와 인덱스를 입력 받도록 정해져 있다는 것을 이를 통해서 알 수 있다.
```

```python
data = [1000, 2000, 3000]
index = ["메로나", "구구콘", "하겐다즈"]  
s = Series(data=data, index=index)  
print(s)        
```

```python
s2 = s.reindex(["메로나", "비비빅", "구구콘"]) 
print(s2)

# None이 나와서 float의 데이터 타입으로 나오게 된다 왜냐? 값을 알 수 없기 때문
# 즉, none 이 나오면 floating 으로 데이터 타입이 바뀐다.
```

```python
s2 = s2.fillna(0) # fillna로 값을 변환하고 변환한 것을 다시 변수에 저장해줘야함!
print(s2)

# fill : 채우겠다  na : none을 (??) : ?의 값으로 채운다 라는 의미이다  
# 즉, 결측치의 값을 채워주는 의미이다.
```

```python
# 이런 데이터 형식을 시계열 데이터라고 한다. 왜냐하면 시간에 따른 변화를 다루는 데이터이기 때문 (날짜 데이터를 다루기 때문ㅇㅇ)

price = [42500, 42550, 41800, 42550, 42650]
data = ["2019-05-31", "2019-05-30","2019-05-29","2019-05-28","2019-05-27"]
s = Series(price, data) # price가 값이 되고 인덱스 위치에 data 가밧을 넣어서 date의 값들이 인덱스가 된다.
print(s)
```

```python
from pandas import Series

data = [1000, 2000, 3000]
index = ["메로나","구구콘","하겐다즈"]
s = Series(data=data, index=index)
s
```

```python
print(s.iloc[0:2])   # i 가 붙은 건 위치 인덱스  / 그냥 loc[location]는 라벨 인덱스인다.
```

```python
print(s.loc["메로나":"구구콘"])
```

```python
indice = [0,2]
print(s.iloc[indice])
# 위치 정보에 변수를 넣어서 위치 정보의 데이터를 변수로 받아들여서 읽는 것
```

```python
print(s.iloc[[0,2]])
# 바로 리스트를 기술했다는 것으로 위에 것과 변수를 사용했나? 아니면 그대로 사용했나의 차이
```

```python
indice = ["메로나","하겐다즈"]
print(s.loc[indice])
print(s.loc[["메로나","하겐다즈"]]) # 여기서는 라벨인덱스 이기 때문에 loc 를 사용한다.
```

```python
s
```

```python
# 값을 수정하는 것 /

s.loc["메로나"] = 500 # s.iloc[0] = 500, s['메로나'] = 500
print(s)
```

```python
s.loc['비비빅'] = 500
print(s)
```

```python
# 내용을 삭제하는 함수로 drop을 사용한다. / series객체에 변화가 발생한다
# 이렇게 변화가 일어나면 변화되지 전의 객체는 삭제가 된다.
# 그렇기에 변수에 저장해야한다.
s1 = s.drop('메로나')
print(s1) # 메로나를 삭제한 객체
print(s)  # 메로나를 삭제하기 전의 객체
```

```python
s = pd.Series([9904312, 3448737, 2890451, 2466052], 
              index=["서울","부산","인천","대구"])
s2 = pd.Series({"서울":9631482, "부산":3393191, "인천":2632035, "대전":1490158},
              index=["부산", "서울", "인천","대전"])
ds = s- s2
print(ds)


# None이 포함된 데이터는 float로 취급 
```

```python
# 이렇게 각각의 데이터에 연산이 일어나는 것을 '브로드캐스팅 연산'이라고 한다.
# 시리즈 데이터에 여러개의 값을 가지고 있고 연산을 하는데 입력된 값은 하나인 경우 
# 브로드캐스팅 연산

s3 = s / 1000000
print(s3)
```

```python
## 동일한 위치에 있는 값만 가져와서 빼주는 것이다. 


print(s.values - s2.values)
```

```python
# notnull() : null이 아닌 경우를 의미
# 데이터를 필터링하는데 사용함
print(ds[ds.notnull()])
```

```python
ds.notnull()
```

```python
rs = (s - s2) / s2 * 100
print(rs)
```

```python
rs = rs[rs.notnull()] # null 값을 제외하기에 공통된 key를 가진 친구들만 연산하게 된다.
print(rs)
```

```python
rs['부산'] = 1.63 # 수정
rs['대구'] = 1.41 # 추가

print(rs)
```

```python
del rs['서울'] # 삭제
print(rs)
```

```python
from pandas import Series

date = ["6/1", "6/2", "6/3", "6/4", "6/5"]
high = Series([42800, 42700, 42050, 42950, 43000], index=date)
low = Series([42150, 42150, 41300, 42150, 42350], index=date)
print(high)
print("-----------------------")
print(low)
```

```python
diff = high - low
print(diff)
```

```python
print(diff.max())
```

```python
print(diff.idxmax()) # idx[index] max 값의 인데긋 위치를 구하는 함수
```

```python
import pandas as pd
import numpy as np # 수치연산을 잘하기 위해서 사용하는 라이브러리 like 행렬 

s = pd.Series(['선화','강호',np.nan,'소정','우영'])
print(s)

# np.nan 은 null data를 나타내는 것이기에 nan을 제외한 4가지의 값만 가지고 있다.
```

```python
s.isnull() # null 로 되어있는 위치를 찾으면 True로 반환
```

```python
s.isna() # na = null로 둘이 동일하다. 즉 isna() = isnull()로 null 값을 반환한다.
```

```python
s[s.isnull()] # null의 개수만 출력
```

```python
s[s.notnull()] # null이 아닌 것만 출력 
```

```python
def remove_comma(x):
    #print(x, 'in function')
    #return x
    return int(x.replace(",","")) # replace(대체 전, 대체 후) 의 값을 넣어서 ,의 값을 공백으로 대체해서 , 를 사라지게 해줌
```

```python
s = Series(["1,234", "5,678", "9,876"])
result = s.map(remove_comma) # map 이 series가 가지고있는 객체의 데이터만큼 반복 호출한다.
print(result)
```

```python
s = Series(["1,234", "5,678", "9,876"])
result = list(s.map(remove_comma)) # series 객체를 list 형태로 반환
print(result)
```

```python
from pandas import Series

data = [3.1, 2.0, 10.1, 5.1]
index = ['000010', '000020', '000030', '000040']
s = Series(data=data, index=index)
print(s)
```

```python
s1 = s.sort_values() 
print(s1)

# 값을 기준으로 오름차순 정렬 , 문자는 유니코드 값으로 정렬 65 A: 97a 이기에 소문자 a가 더큼
```

```python
s1 = s.sort_values(ascending=False) 
print(s1)

# 값을 기준으로 오름차순 정렬 , 문자는 유니코드 값으로 정렬 65 A: 97a 이기에 소문자 a가 더큼
```

```python
s2 =s.rank()
print(s2)
```

```python
s2 =s.rank(ascending=True) # rank() = rank(ascending=True) 가 디폴트
print(s2)
```

```python
s2 =s.rank(ascending=False) # False를 주면 내림차순 순서
print(s2)
```
