```python
# 데이터플레임이나 시리즈에서 원하는 데이털르 선택(인덱싱)하는 방법 = loc, iloc
# loc = 라벨 기반 인덱승 



import pandas as pd
import numpy as np
# 값, 인덱스, 컬럼 순으로 적는다
df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], index=['row1','row2','row3'],
                 columns=['col1','col2','col3'])
df
```

```python
# loc가 라벨 기반 인덱싱이다 즉, **행 이름(index)**을 기준으로 데이터를 선택하라는 뜻
# 여기서는 row1과 row2의 행을 인덱싱해서 출력해라는 뜻으로 결과값이 row1과 row2만 나온다


result = df.loc[['row1','row2']]
result
```

```python
# DataFrame df에서 행 인덱스가 row2dlrh 컬럼 인덱스가 col2인 위치의 값을 추출하여 result에 저장한다
# pandas는 내부적으로 Numpy 배열을 사용해서 데이터를 저장하기에 일반 int가 아니라 Numpy의 정수타입으로 반환된다.

result = df.loc['row2','col2']
result
```

```python
# row1부터 row3까지의 행에서 col2 열의 값만 가져와라 
# 이때 이 값은 series값으로 반환된

## series 형태로 반환된 이유는 '선택한 차원이 1개만 남았기 때문'
## 선택결과가 2차원이면 = DataFrame
## 행또는 열 = Series
## 하나의 값 = scalar(값)
# 의 형태 규칙을 띄고 있다.

result = df.loc['row1' : 'row3', 'col2']
result
```

```python
# dataframe 에서 행을 선택할지 말지의 여부를 의미한다.
# 현재 f,t,f 이기 때문에 0번 1번 ~~~~순으로 행이 순서가 진행되기 때문에 두번째는 1번 행이다
# 그렇기에 1번째 행만 선택한다는 의미를 bool 이 담고 있고 이를 loc에 집어넣으면
# 1번 인덱스가 도출된다



bool2 = [False, True, False]
result = df.loc[bool2]
result
```

```python
# col3에 대해서 값이 5보다 큰 값에 대해서만 출력을 해주는 

# df.loc[df['col3'] > 5] 는 조건식이다
# 의미 : c0l3의 열을 선택하고 이 선택한 열의 값 중 5의 값보다 큰것을
# 선택한다(loc). 는 의미이다. # 즉 5보다 큰 값을 true로 반환하고 그 트루에 해당하는 것만 반환

result = df.loc[df['col3'] > 5]
result
```

```python
# 행, 열 상관없이 리스트형태로 만들어주면 데이터 프레임 형태로 반환된다.


# loc [행, 열] 의 순서로 입력을 하고 입력을 받는다. 그렇기에 현재 코드를 분서해보면
# col3 > 5 은 col3 에서 5보다 큰것을 True로 반환하고 거기서 True인 것에 한해서
# col2인 것을 반환해서 result에 저장하라는의미 


result = df.loc[df['col3']>5,['col2']]
result
```

```python
# df을 전달받아서 col2가  5와 같은 데이터 플레임의 행을 추출한다

# df['col2']==5인 값을 찾아서 lambda라는 함수에 넣고 그에 해당된 행을 선택[loc]해서
# result 변수에 넣어서 출력해라는 뜻

result = df.loc[lambda df : df['col2'] == 5]
result
```

```python
# slice(None) : [:]슬라이스와 동일한 의미, 멀티 인덱스의 특징 레벨 내의 모든 값


# 데이터 전체를 선택할 때 보통 [:]의 슬라이스를 사용하게 되는데 열을 1층 2층ㅇ로 나눈
# 멀티 인덱스 구조에서는 튜플 형태인 () 안에 묶어서 전달해야한다.
# 그런데 슬라이스를 튜플 안에 넣을 수가 없다.

# 즉, 멀티인덱스이고, 튜플 형태로 전달해야하고, 슬라이스(:)를 튜플안에 전달할 수 없으니
# slice(None)이라는 함수를 사용해서 이전의 모든 값을 불러와준다.


## 해석
# loc를 사용했으니 행의 순서대로 불러오고 ==> 2020/06월의 행인 것
# 1층 상위 레벨의 열은 silce(None)을 사용해 모든 1층 상위에 해당하는 것 중 "컨센서스"에
# 해당하는 모든 값을 출력해라 
# but, list 형태로 출력하지 않았기네 series 형태로 출력해라 
## if datafream 형태로 출력하고 싶다면 리스트 형태로 [행,열]을 묶어서 작성해라
# 는 의미이다.

df.loc["2020/06", (slice(None), "컨센서스")]
```

```python
# 위에서 작성한 것을 dataframe 형태로 출력하기 위해 list로 행을 묶어서 출력함

df.loc[["2020/06"], (slice(None), "컨센서스")]
```

```python
# iteems를 사용하면 라벨명을 가지고 오면 해당 데이터를 필터링 할 수 있다.

# but df.list해서 원래 하던대로 출력 할 수도 있다. 이것은 선택의 문제이다.

df.filter(items=['alpha','delta'])
```

```python
# df의 열 이름 중에 ta글자가 포함된 열만 다 골라줘라는 뜻

df.filter(like='ta')
```

```python
# mn을 다 포함하고 있는 라벨문을 모두 출력하라는 의미이다.
# 즉, 데이터프레임의 열 이름 중 소문자 m,n이 하나라고 포함된 열만 다 골라줘라는 뜻
### regex 는 정규 표현식의 의미로 해석해서 출력하라는 뜻


df.filter(regex='[mn]')
```

```python
# ^ 는 시작의 의미 

# 즉 g로 시작하는 모든 행에 해당하는 것을 필터링해서 출력하라는 의미

df.filter(regex='^g')
```

```python
# $ 는 끝의 의미

# 즉 $로 끝나는 모든 행읠 필터링 해서 출력하라는 의미

df.filter(regex='a$')
```
