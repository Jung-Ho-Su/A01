```python
# 문제3 / 다음과 같은 Seriees를 생성하는 코드를 작성하시오.
from pandas import Series
data = ['apple', np.nan, 'banana', 'kiwi', 'gubong']
index = ['가','나','다','라','마',]

s = Series(data=data, index=index)
print(s)
```

```python
# 문제 4 / 위 예제에서 결측치가 아닌 데이터만 필터링하여 출력하는 코드를 작성하시오.
s[s.notnull()]
```

```python
# 문제 5 / 위 예제에서 결측치 데이터만 필터링하여 출력하는 코드를 작성하시오.
s[s.isnull()]
```

```python
# 문제 6

# seed는 난수를 발생시킬 때 설정해주면 매번 동일한 난수를 만들어 줄 수 있다.
np.random.seed(0) 
sample = pd.Series(np.random.randint(100, 200, size=(10,)))
sample[2:7]
```

```python
# 문제 7

np.random.seed(0)
sample2 = pd.Series(np.random.randint(100, 200, size=(10,)), index=list('가나다라마바사아자차'))

print(sample2['바':'차'], end='\n\n') # 현재는 라벨 인덱스를 사용했는데, 위치 인덱스를 사용해서도 표현할 수 있다 즉, 숫자를 사용해서 표현 가
print(sample2['가':'다'], end='\n\n')
print(sample2['나':'바'], end='\n\n')
```
