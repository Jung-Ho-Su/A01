```python
import pandas as pd

s1 = pd.Series([0, 1], index=['A', 'B'])
s1
```

```python
s2 = pd.Series([2,3,4], index=['A','B','C'])
s2
```

```python
s3 = pd.concat([s1,s2]) # concat는 단순 연결이라서 인덱스를 새롭게 설정하는 작업인 reset_index가 필요하다.
s3
```

```python
s3 = pd.concat([s1, s2], axis=1)
s3
```

```python
import numpy as np
df1 = pd.DataFrame(np.arange(6).reshape(3,2),                # arange는 파이선이 제공하는 rang와 같다 # 6 이전까지의 숫자
                  index=['a','b','c'],                       # reshapge(3,2) 3행 2열 구조로 바꿔라!
                  columns=['데이터1','데이터2'])
df1
```

```python
df2 = pd.DataFrame(5 + np.arange(4).reshape(2,2),
                  index=['a','c'],
                  columns=['데이터3','데이터4'])
df2
```

```python
df3 = pd.concat([df1, df2], axis=1) # axis = 1 인 경우 
df3
```

```python
df3 = pd.concat([df1, df2], axis=0) # axis = 0 인 경우 
df3 
```

```python
# join

df1 = pd.DataFrame([[1., 2.],[3., 4.], [5., 6.]],
                  index = ['a', 'c', 'e'],
                  columns=['서울', '부산'])
df1
```

```python
df2 = pd.DataFrame([[7., 8.],[9., 10.],[11., 12.],[13, 14]],
                  index=['b','c','d','e'],
                  columns=['대구','광주'])
df2
```

```python
# join 같은 경우에는 index를 기준으로 병합한다.

df3 = df1.join(df2, how='outer') # 기준이 없으면 index를 기준으로 병합한다.
df3
```

```python
df4 = df1.join(df2, how='inner') # 기준이 없으면 index를 기준으로 병합한다.
df4
```

```python
df5 = df1.join(df2, how='left')
df5
```

```python
import numpy as np

n = np.nan
idx = ['row1', 'row2', 'row3']
col = ['col1', 'col2', 'col3']
data=[[1,2,200],[100,5,6],[7,300,n]]
df = pd.DataFrame(data, idx, col)
df
```

# 시계열 데이타 # 시계열 데이터 # To Data

```python
date_str = ['2018, 1, 1', '2018, 1, 4', '2018, 1, 5'] 
idx = pd.to_datetime(date_str)
idx
```

```python
np.random.seed(0)
s = pd.Series(np.random.randn(3), index=idx)
s
```

```python
pd.date_range('2020-7-1','2020-7-30') # date_range 는 범위를 표시해주는 함수 # freq = D는 Day의 의미
# 실제로 date_range가 많이 쓰인다.
```

```python
pd.date_range(start='2020-7-1', end='2020-7-30') # 시작일과 마지막일을 사용
```

```python
pd.date_range(start='2020-7-1', periods = 7) # 시작일과 기간을 사용
```

```python
pd.date_range(start='2020-7-1', periods = 7, freq ='s')  # 초단위로 변경해서 출력함
```

```python
pd.date_range('2026-1-1', '2026-12-31', freq='W') # 주 단위 # 주 단위를 설정하지 않으면 W-SUN 이 디폴트이다.
```

```python
pd.date_range('2026-1-1', '2026-12-31', freq='W-MON') # 월요일 기준으로 함
```

```python
pd.date_range('2026-1-1', '2026-12-31', freq='W-FRI') # 금요일 기준으로 함
```

```python
pd.date_range('2026-1-1', '2026-12-31', freq="B") # 평일을 의미한다.
```

```python
pd.date_range('2021-1-1','2021-12-31', freq='MS')
```

```python
pd.date_range('2021-1-1','2021-12-31', freq='ME')
```

```python
pd.date_range('2021-1-1','2021-12-31', freq='WOM-2SUN')
```

```python
pd.date_range('2021-1-1','2021-12-31', freq='QE-JAN')
```

# 예제 파일

```python
sample1 = pd.read_csv('sample1.csv') # read_csv(파일 이름)
sample1

# sample1 파일을 그냥 불러옴 
```

```python
sample2 = pd.read_csv('sample2.csv', names=['c1','c2','c3'])
sample2

# sample2 의 컬럼 네임을 c1c2c3로 지정해줌 
```

```python
sample1 = pd.read_csv('sample1.csv', index_col='c1') # index col을 c1으로 지정
sample1

# index 컬럼을 c1으로 지정해서 c1 에 인덱스 컬럼으로 표시되게 함
```

```python
sample3 = pd.read_table('sample3.txt', sep='\t') 
sample3

# 현재 파일은 탭 키를 기준으로 구분되어 있어서 백슬레시로 구분함 
# table, csv 둘 다 사용해도 불러 올 수 있음
```

```python
sample4 = pd.read_csv('sample4.txt', skiprows=[0,1]) # skiprows = [스킾할 행 기술]
sample4

# sample4를 보면 0번째 행과 1번째 행에 불필요한 데이터가 있다 이를 skip하기 위해
# skiprows를 사용해서 0, 1 번 인덱스를 선택해서 skip 했다.
```

```python
# 파일에 누락으로 뜨는 데이터를 nan으로 처리하라는 뜻 : na_values=['누락']
# sample5 에 있는 '누락' 이라는 데이터를 Nan으로 처리해서 결과 값을 보면 Nan으로 뜨게 된다.

sample5 = pd.read_csv('sample5.csv', na_values=['누락'])
sample5
```
