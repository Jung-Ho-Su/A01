### pivot_table
형식) pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, margins_name='All')
- data: 분석할 데이터프레임 (메서드일 때는 필요하지 않음)
- values: 분석할 데이터프레임에서 분석할 열
- index: 행 인덱스로 들어갈 키 열 또는 키 열의 리스트
- columns: 열 인덱스로 들어갈 키 열 또는 키 열의 리스트
- aggfunc: 분석 메서드 지정
- fill_value: NaN 대체 값 설정
- margins: 오른쪽과 아래에 합계를 붙일지 여부 지정
- margins_name: 합계 열(행)의 이름 지정

```python
df2 = pd.DataFrame({'key1':['A','A','B','B','A'],
                   'key2' : ['one','two','one', 'two', 'one'],
                   'data1':[1,2,3,4,5],
                   'data2':[10,20,30,40,50]})
df2
```

```python
#                데이타,  인덱스, 컬럼
df2.pivot_table('data1','key1','key2')

# aggfunc='mean' 현재 디폴트이다.
# 여기서 집계함수가 빠져있다. # 여기서는 디폴드 값으로 평균을 집계함수로 사용하고 있다.
```

```python
# 위에랑 똑같다.
df2.pivot_table('data1','key1','key2', aggfunc='sum')
```

```python
# 위에랑 똑같다.
df2.pivot_table('data1','key1','key2', aggfunc=sum)
```

```python
# margins 비벗 테이블의 총 합계 평균 # 근데 디폴트가 all 로 나타나서 잘 알 수가 없다.

df2.pivot_table('data1', 'key1', 'key2', margins=True)
```

```python
# margin은 총계의 평균의 의미, 여기서 디폴트가 all이라 헷갈리지 않게 
# margins_name = "평균"으로 잡아준다.

df2.pivot_table('data1', 'key1', 'key2', margins=True, margins_name='평균')
```

```python
# key1, key2를 기준으로 하위 그룹을 묶어 값을의 평균을 구한다.
df2.pivot_table('data1', index=['key1','key2'])
```
