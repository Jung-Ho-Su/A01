```python
# 문제 / 아래 코드를 참고하여 주어진 문제에 대한 코드를 완성하시오.


### (1) iris 데이터에서 species별 sepal_length의 평균을 구한 후 이를 'bar' 타입으로 그려보세요.
### (2) iris 데이터에서 품좀별 데이터의 개수를 파이 차트 형태로 그려보세요.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family']='Malgun Gothic' # 글꼴을 맑은 고딕을 사용하겠다는 의미
plt.rcParams['axes.unicode_minus']=False # 마이너스 부호가 깨지는 현상을 막아줌
iris = sns.load_dataset("iris")
iris
```

```python
# (1) iris 데이터에서 species별 sepal_length의 평균을 구한 후,
# 이를 'bar' 타입으로 그려보세요.
df = iris.groupby(iris.species).mean()
df['sepal_length'].plot.bar(rot=0)
plt.title("각 종의 Sepal Length 평균")
plt.xlabel("종 (Species)")
plt.ylabel("Sepal Length 평균")
plt.ylim(0, 10)
plt.show()
```

```python
# (2) iris 데이터에서 품좀별 데이터의 개수를 파이 차트 형태로 그려보세요.
import seaborn as sns
import matplotlib.pyplot as plt

df2 = iris['species'].value_counts()
df2.plot.pie(autopct='%2.2f%%')
plt.title("품종별 데이터 개수")
plt.ylabel("")
plt.show()
```
