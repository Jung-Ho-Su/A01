```python
import matplotlib.pyplot as plt # Matplotlib의 시각화 모듈을 불러와서 plt라는 별칭으로 사용
import numpy as np               # 수치 계산에 최적화된 Numpy 라이브러리를 불러와서 np로 사용

# np.arange(시작, 끝, 간격) 함수:
# -4.5부터 시작하여 5 미만까지 0.5 간격으로 숫자를 생성하여 배열(Array)을 만듭니다.
# x축의 정밀도를 결정하며, 간격(0.5)이 작을수록 그래프의 곡선이 부드러워집니다.
x = np.arange(-4.5, 5, 0.5)     

# y = 2 * (x의 제곱) 연산:
# Numpy 배열인 x의 모든 원소에 대해 각각 제곱을 하고 2를 곱합니다.
# 이 코드는 포물선 형태를 그리는 전형적인 '2차 함수' 데이터를 생성합니다.
y = 2 * x**2                     

# plt.plot(x, y) 함수:
# 생성된 x와 y 좌표들을 선으로 연결하여 그래프를 그립니다.
# 별도의 스타일 지정이 없으므로 기본값인 '파란색 실선'으로 출력됩니다.
# plot 함수는 주로 plot(x, y)의 형태로 x 축과 y 축 데이터를 입력받아 점과 선을 좌표계에 그린다.
plt.plot(x, y)                   

# plt.show() 함수:
# 메모리에 생성된 그래프 객체를 화면(Window)에 최종적으로 띄워줍니다.
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 도구인 pyplot을 plt라는 별칭으로 가져오기
import numpy as np               # 수치 계산 배열 처리를 위한 numpy를 np라는 별칭으로 가져오기

# x축 데이터 생성: -4.5부터 5 미만까지 0.5 간격으로 숫자를 배열함
# 그래프의 가로 범위를 결정하며, 간격이 촘촘할수록 곡선이 부드럽게 그려짐
x = np.arange(-4.5, 5, 0.5)     

# y축 데이터 생성 (서로 다른 3개의 함수 식 정의)
y1 = 2*x**2                        # 2차 함수: 아래로 볼록한 포물선 형태 (원점 근처)
y2 = 5*x + 30                      # 1차 함수: 기울기가 5이고 y절편이 30인 직선 형태
y3 = 4*x**2 + 10                   # 2차 함수: y1보다 폭이 좁고 위로 10만큼 평행이동한 포물선

# plt.plot() 함수를 연속해서 호출하면 하나의 Axes(차트 영역) 안에 그래프가 겹쳐서 그려짐
# Matplotlib은 내부적으로 피규어(도화지)와 액시스(축 영역) 객체를 생성하고 관리함
plt.plot(x, y1)  # 첫 번째 데이터(y1)를 그림 (기본색: 파란색)
plt.plot(x, y2)  # 두 번째 데이터(y2)를 겹쳐서 그림 (기본색: 주황색)
plt.plot(x, y3)  # 세 번째 데이터(y3)를 겹쳐서 그림 (기본색: 초록색)

# plt.show(): 메모리에 쌓인 모든 그래프 요소를 최종적으로 윈도우 창에 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 라이브러리를 불러와 np로 별칭 지정

x = np.arange(-5, 5, 0.1)        # -5부터 5 미만까지 0.1 간격으로 x축 데이터 생성
y1 = x**2 -2                     # 2차 함수 데이터 생성 (포물선)
y2 = 20*np.cos(x)**2             # 삼각함수 데이터 생성 (코사인 제곱 형태)

# --- 피규어 객체 생성 및 선택 과정 이해 ---

plt.figure(1)                    # 1번 피규어(도화지)를 생성함. 이제부터 모든 작업은 1번에 기록됨.
plt.plot(x, y1)                  # 현재 활성화된 1번 피규어에 y1(2차 함수) 그래프를 그림

plt.figure(2)                    # 2번 피규어를 새롭게 생성함. 이제 작업 대상이 2번으로 바뀜.
plt.plot(x, y2)                  # 현재 활성화된 2번 피규어에 y2(삼각함수) 그래프를 그림

plt.figure(1)                    # 기존에 있던 1번 피규어를 다시 선택함. 작업 대상이 다시 1번이 됨.
plt.plot(x,y2)                   # 1번 피규어에 y2 그래프를 추가함 (y1과 y2가 한 창에 겹쳐짐)

plt.plot(2)                      # [참고] 이 코드는 y축 값으로 '2'라는 단일 값을 주어 점을 찍으려 시도함
                                 # (하지만 x축 데이터가 없어 그래프가 제대로 보이지 않거나 에러 가능성 있음)

plt.clf()                        # Clear Figure: 현재 선택된 1번 피규어의 모든 내용(y1, y2)을 싹 지움
plt.plot(x, y1)                  # 깨끗해진 1번 피규어에 다시 y1(2차 함수)만 새로 그림


plt.show()                       # 최종적으로 생성된 모든 피규어 창(1번과 2번)을 화면에 출력함
```

# 라인 플롯


```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정

# 한글 깨짐 방지를 위해 시스템에 설치된 '맑은 고딕' 폰트를 기본 폰트로 설정
plt.rcParams["font.family"] = 'Malgun Gothic'

# 그래프 상단에 표시될 제목 설정
plt.title("라인 플롯(line plot)")

# plt.plot() 함수 상세 옵션 해석:
# [10, 20, 30, 40] : x축 데이터 좌표
# [1, 4, 9, 16]    : y축 데이터 좌표
# c='b'            : color, 선의 색상을 파란색(blue)으로 설정
# lw=5             : linewidth, 선의 두께를 5로 설정
# ls='--'          : linestyle, 선의 모양을 파선(dashed)으로 설정
# marker='o'       : 마커 종류를 원형(circle)으로 설정
# ms=15            : markersize, 마커의 전체 크기를 15로 설정
# mec='g'          : markeredgecolor, 마커의 테두리 색상을 녹색(green)으로 설정
# mew=5            : markeredgewidth, 마커 테두리의 두께를 5로 설정
# mfc='r'          : markerfacecolor, 마커 내부의 채우기 색상을 빨간색(red)으로 설정
plt.plot([10, 20, 30, 40],[1, 4, 9, 16], c='b', lw=5, ls='--', marker='o', ms=15,
         mec='g', mew=5, mfc='r')

# plt.xlim(최소, 최대) : x축이 보여지는 범위(축 한계)를 0부터 50까지로 제한
plt.xlim(0, 50)

# plt.ylim(최소, 최대) : y축이 보여지는 범위(축 한계)를 0부터 20까지로 제한
plt.ylim(0, 20)

# 설정한 모든 스타일이 적용된 그래프를 화면에 출력
plt.show()
```

```python
from matplotlib import font_manager # 시스템에 설치된 폰트 파일을 관리하는 모듈 불러오기
import matplotlib.pyplot as plt      # 시각화 라이브러리를 불러와 plt로 별칭 지정

# 1. 폰트 경로 지정: 윈도우 시스템 폴더 내의 '맑은 고딕(malgun.ttf)' 파일 경로 설정
font_location = 'C:\Windows\Fonts\malgun.ttf'

# 2. 폰트 이름 추출: 지정한 경로의 폰트 파일로부터 실제 폰트의 이름(예: 'Malgun Gothic')을 가져옴
font_name = font_manager.FontProperties(fname=font_location).get_name()

# 3. 폰트 스타일 사전(Dictionary) 생성:
# 'family': 추출한 폰트 이름 적용, 'color': 글자색 검정, 'size': 글자 크기 20
font = {'family':font_name, 'color':'black', 'size':20}

# 4. 제목 설정: fontdict 인자에 위에서 만든 스타일 사전을 전달하여 제목 꾸미기
# fontdict 글꼴 스타일을 딕셔너리 형태로 한꺼번에 지정할 때 사용하는 매개변수
plt.title('라인 플롯(line plot)', fontdict=font)

# 5. 그래프 그리기:
# x=[10, 20, 30, 40], y=[1, 4, 9, 16] 데이터 사용
# c='b': 파란색 선, lw=5: 선 두께 5, ls='--': 점선 스타일
# marker='o': 원형 점, ms=15: 점 크기 15, mec='g': 점 테두리 초록, mew=5: 테두리 두께 5, mfc='r': 점 내부 빨강
plt.plot([10, 20, 30, 40],[1, 4, 9, 16], c='b', lw=5, ls='--', marker='o', ms=15,
         mec='g', mew=5, mfc='r')

# 6. x축 출력 범위 설정: 0부터 50까지 보이도록 제한
plt.xlim(0, 50)

# 7. y축 출력 범위 설정: 0부터 20까지 보이도록 제한
plt.ylim(0, 20)

# 8. 최종 그래프 출력
plt.show()
```

```python
from matplotlib import font_manager  # 폰트 파일을 관리하는 모듈 불러오기
import matplotlib.pyplot as plt      # 시각화 라이브러리를 불러와 plt로 별칭 지정

# 1. 폰트 경로 지정: 윈도우 시스템의 '맑은 고딕' 폰트 파일 위치 설정
font_location = 'C:\Windows\Fonts\malgun.ttf'
# 2. 폰트 이름 추출: 파일 경로로부터 실제 폰트 패밀리 이름(예: 'Malgun Gothic')을 가져옴
font_name = font_manager.FontProperties(fname=font_location).get_name()

# 3. rc 파라미터 설정: 
# plt.rc 함수를 사용하여 그래프의 기본 폰트를 추출한 폰트 이름으로 전역 설정함
# 이제 제목뿐만 아니라 축 레이블, 범례 등 모든 텍스트에 이 폰트가 적용됨
plt.rc('font', family=font_name) 

# 4. 제목 설정: 
# [주의] 이전에 정의한 'font' 딕셔너리가 없다면 fontdict=font는 에러가 날 수 있음
plt.title('라인 플롯(line plot)')

# 5. 그래프 그리기:
# x=[10, 20, 30, 40], y=[1, 4, 9, 16] 데이터 사용
# 스타일: 파란색 점선, 두꺼운 선(5), 녹색 테두리의 빨간색 원형 마커(15)
plt.plot([10, 20, 30, 40],[1, 4, 9, 16], c='b', lw=5, ls='--', marker='o', ms=15,
         mec='g', mew=5, mfc='r')

# 6. x축 및 y축 표시 범위 설정
plt.xlim(0, 50) # x축을 0~50 범위로 고정
plt.ylim(0, 20) # y축을 0~20 범위로 고정

# 7. 그래프 저장하기:
# 현재까지 그려진 피규어 객체를 'fig01.png'라는 이름의 파일로 저장함
# dpi=100 : 해상도(Dots Per Inch)를 100으로 설정 (숫자가 높을수록 고화질)
plt.savefig('fig01.png', dpi=100)

# 8. 최종 그래프 화면 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 라이브러리를 불러와 np로 별칭 지정

# np.linspace(시작, 끝, 개수) 함수:
# -pi(-3.14...)부터 pi(3.14...)까지의 구간을 정확히 256등분하여 숫자를 생성함
# 데이터 포인트가 많을수록 코사인 곡선이 아주 부드럽게 그려짐
## linspace(linear space)는 시작점과 끝점 사이를 지정된 개수만큼 
## 균일한 간격으로 나누어 배열(ndarray)을 생성하는 파이썬 NumPy 및 MATLAB의 핵심 함수
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)                    # 생성된 x값들에 대한 코사인(cos) 값 계산

# 그래프 상단 제목 설정 (LaTeX와 문자열 눈금 정의에 관한 예제임을 명시)
plt.title('LaTex, 문자열로 tick label 정의')

# 계산된 x, c 데이터를 사용하여 기본 선 그래프 생성
plt.plot(x, c)

# plt.xticks([위치 리스트], [라벨 리스트]):
# 1. 위치 리스트: x축의 어느 좌표에 눈금을 표시할지 정함 (-pi, -pi/2, 0, pi/2, pi)
# 2. 라벨 리스트: 해당 위치에 숫자 대신 보여줄 글자를 정함
###    - r'$...$' : LaTeX 문법을 사용하여 그리스 문자 파이(pi)를 수학 기호로 출력함
#    - r은 Raw string으로, 백슬래시(\)를 특수문자로 인식하지 않게 함
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# plt.yticks([위치 리스트], [라벨 리스트]):
# y축의 -1, 0, 1 위치에 숫자 대신 'Low', 'Zero', 'High'라는 문자열을 입힘
plt.yticks([-1, 0 ,1], ['Low', 'Zero', 'High'])

# 그래프 바탕에 격자(Grid)를 표시하여 눈금 위치를 읽기 쉽게 만듦
plt.grid(True)

# 최종 그래프 출력 (참고: plt.show() 뒤에 괄호를 붙여야 정상 실행됩니다)
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 라이브러리를 불러와 np로 별칭 지정

# 0부터 2 미만까지 0.2 간격으로 x축 데이터 생성 (0, 0.2, 0.4, ..., 1.8)
x = np.arange(0, 2, 0.2)

# 1. 첫 번째 그래프: y = x (일차 함수)
# 'bo' : Blue(파란색) 색상에 'o'(원형) 마커를 사용 (선 없이 점만 찍힘)
plt.plot(x, x, 'bo')

# 2. 두 번째 그래프: y = x^2 (이차 함수)
# color='#e35f62' : HEX 코드를 사용해 붉은 계열 색상 지정
# marker='*'      : 마커 모양을 별표(*)로 설정
# linewidth=2     : 연결선의 두께를 2로 설정
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)

# 3. 세 번째 그래프: y = x^3 (삼차 함수)
# color='springgreen' : 미리 정의된 색상 이름 사용 
# (주의: 색상 이름 앞에 #을 붙이면 에러가 날 수 있으므로 실제 실행 시에는 'springgreen'으로 수정 권장)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)

# plt.xticks(위치, 라벨): x축의 숫자 눈금을 문자열(월 이름)로 변경
# 0부터 2.2까지 0.2 간격 위치에 'Jan'부터 'July'까지 순서대로 입힘
# 빈 문자열('')을 넣어 중간 눈금은 글자 없이 표시함
plt.xticks(np.arange(0, 2.2, 0.2), labels=['Jan', '', 'Feb','','Mar','','May','','June','','July'])

# plt.yticks(위치, 라벨): y축의 0부터 6까지 위치에 '0'부터 '6GB'까지 단위 이름을 입힘
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

# 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 라이브러리를 불러와 np로 별칭 지정

# 0부터 2 미만까지 0.2 간격으로 x축 데이터 생성 (0, 0.2, 0.4, ..., 1.8)
x = np.arange(0, 2, 0.2)

# 1. 세 가지 서로 다른 데이터 세트(x, x^2, x^3)를 그래프에 그림
plt.plot(x, x, 'bo') # 파란색(b) 원형(o) 마커로 점만 표시
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2) # 빨간색 별 마커와 두께 2의 선
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9) # 연두색 삼각형 마커와 크기 9 설정

# 2. x축과 y축의 눈금 위치(ticks)와 표시될 이름(labels)을 수동으로 설정
plt.xticks(np.arange(0, 2.2, 0.2), labels=['Jan', '', 'Feb','','Mar','','May','','June','','July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

# 3. x축 눈금(Tick)의 상세 스타일 설정
# axis = 'x'       : x축을 대상으로 설정함
# direction = 'in' : 눈금 선을 그래프 안쪽 방향으로 향하게 함
# length = 3       : 눈금 선의 길이를 3으로 설정
# pad = 6          : 눈금 선과 라벨(글자) 사이의 간격을 6으로 설정
# labelsize = 14   : 눈금 라벨(월 이름)의 글자 크기를 14로 설정
# labelcolor = 'green' : 눈금 라벨의 글자 색상을 초록색으로 설정
# top = True       : 그래프 상단 테두리에도 눈금 선을 표시함
plt.tick_params(axis = 'x', direction = 'in', length = 3, pad = 6, labelsize = 14,
                labelcolor = 'green', top = True)

# 4. y축 눈금(Tick)의 상세 스타일 설정
# axis = 'y'            : y축을 대상으로 설정함
# direction = 'inout'   : 눈금 선이 그래프 안팎으로 걸쳐지게 표시함
# length = 10           : 눈금 선의 길이를 10으로 설정 (x축보다 길게 강조)
# pad = 15              : 눈금 선과 라벨(0, 1GB 등) 사이의 간격을 15로 설정
# labelsize = 12        : 눈금 라벨의 글자 크기를 12로 설정
# width = 2             : 눈금 선 자체의 굵기를 2로 설정
# color = 'r'           : 눈금 선의 색상을 빨간색(red)으로 설정
plt.tick_params(axis = 'y', direction = 'inout', length = 10, pad = 15, labelsize = 12,
                width = 2 , color = 'r')

# 5. 최종 그래프 출력
plt.show()

# [참고] plt.show()가 두 번 호출되었으나, 첫 번째 호출에서 창을 닫으면 두 번째 호출은 빈 창이 뜨거나 무시됩니다.
plt.show()
```

tick_prams() 함수 : 눈금의 스타일을 다양하게 설정
- axis: 설정이 적용될 축을 지정
- direction : 'in','out'으로 설정하면 눈금이 안/밖으로 표시, {'in', 'out', 'inout'} 중 선택
- length : 눈금의길이를 지정
- pad : 눈금과 레이블과의 거리를 지정
- labelsize : 레이블의크기를지정
- labelcolor : 레이블의 색상을지정
- top/bottom/left/right : True/False로 지정하면 눈금이 표시될 위치를 선택
- width : 눈굼의 너비를 지정
- color : 눈금의 색상을 지정

```python
import matplotlib.pyplot as plt  # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 라이브러리를 불러와 np로 별칭 지정

# 0.0부터 5.0 미만까지 0.2 간격으로 x축 데이터 생성 (시간 또는 독립 변수 t)
t = np.arange(0., 5., 0.2)

# 그래프 상단의 메인 제목 설정
plt.title('Line Plot')

# plt.plot() 함수 하나에 여러 개의 (x, y, 스타일) 세트를 쉼표로 연결하여 그리기:
# 1. t, t, 'r--'          : y=t 선을 빨간색(red) 파선(dashed)으로 그림
# 2. t, 0.5 * t**2, 'bs:' : y=0.5*t^2 선을 파란색(blue) 사각형(square) 마커와 점선(dotted)으로 그림
# 3. t, 0.2 * t**3, 'g^-' : y=0.2*t^3 선을 초록색(green) 삼각형(triangle_up) 마커와 실선(solid)으로 그림
plt.plot(t, t, 'r--', t, 0.5 *t**2, 'bs:', t, 0.2 * t**3, 'g^-')

# plt.legend() 범례 설정:
# ['Red', 'Blue', 'Green'] : 각 선에 대응하는 이름표 설정
# loc = 'best'             : 데이터 흐름을 방해하지 않는 최적의 위치를 Matplotlib이 자동으로 찾아 배치함
# ncol = 3                 : 범례 목록을 세로가 아닌 가로 3열(1행 3열)로 나열함
# fontsize = 7             : 범례 글자 크기를 7로 작게 설정하여 그래프 영역을 확보함
plt.legend(['Red', 'Blue', 'Green'], loc = 'best', ncol=3, fontsize = 7) 

# 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 라이브러리를 불러와 np로 별칭 지정

# 0.0부터 5.0 미만까지 0.2 간격으로 x축 데이터 생성
t = np.arange(0., 5., 0.2)

# 그래프 상단의 제목 설정
plt.title('Line Plot')

# 한 줄의 plot 함수에 세 가지 데이터를 스타일과 함께 그리기:
# 1. t, t, 'r--'          : 빨간색 파선
# 2. t, 0.5 * t**2, 'bs:' : 파란색 사각형 마커와 점선
# 3. t, 0.2 * t**3, 'g^-' : 초록색 삼각형 마커와 실선
plt.plot(t, t, 'r--', t, 0.5 *t**2, 'bs:', t, 0.2 * t**3, 'g^-')

# plt.legend() 범례의 세부 설정:
# ['Red', 'Blue', 'Green'] : 각 선의 이름표(라벨) 정의
# loc = 'best'             : 그래프의 데이터와 겹치지 않는 가장 적절한 위치에 자동 배치
# ncol = 3                 : 범례 아이템을 가로로 3열(한 줄)로 나열함
# fontsize = False         : 범례 글자 크기 설정 (일반적으로 숫자나 'medium' 등을 사용함)
# shadow = True            : 범례 박스 뒤에 입체감을 주는 '그림자 효과'를 추가함
plt.legend(['Red', 'Blue', 'Green'], loc = 'best', ncol=3, fontsize = False, shadow= True) 

# 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 라이브러리를 불러와 np로 별칭 지정

# 0.0부터 5.0 미만까지 0.2 간격으로 x축 데이터 생성 (독립 변수 t)
t = np.arange(0., 5., 0.2)

# 그래프 상단의 메인 제목 설정
plt.title('Line Plot')

# 각 plt.plot() 호출은 하나의 선(Line)을 차트 영역(Axes)에 추가합니다.
# 1. 첫 번째 선: y = t (일차 함수)를 빨간색(r) 파선(--)으로 그림
plt.plot(t, t, 'r--') 

# 2. 두 번째 선: y = 0.5 * t^2 (이차 함수)를 파란색(b) 사각형(s) 마커와 점선(:)으로 그림
plt.plot(t, 0.5 *t**2, 'bs:') 

# 3. 세 번째 선: y = 0.2 * t^3 (삼차 함수)를 초록색(g) 삼각형(^) 마커와 실선(-)으로 그림
plt.plot(t, 0.2 * t**3, 'g^-') 


# plt.legend() 범례 설정:
# ['Red', 'Blue', 'Green'] : 각 선에 대응하는 이름표(Label) 설정
# loc = 'best'             : 데이터와 겹치지 않는 최적의 위치를 자동 탐색하여 배치
# ncol = 3                 : 범례 항목을 가로로 3개 나열 (한 줄로 배치)
# fontsize = False         : 범례 글자 크기 (기본 설정을 유지하거나 특정 숫자 지정 가능)
# shadow = True            : 범례 박스 뒤에 그림자 효과를 주어 입체감 부여
plt.legend(['Red', 'Blue', 'Green'], 
           loc = 'best', ncol=3, fontsize = False, shadow= True) 

# 최종적으로 생성된 피규어 객체를 화면에 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # Matplotlib의 시각화 도구인 pyplot 모듈을 가져와 plt라는 별칭으로 사용 (가장 핵심적인 함수들이 들어있음)
import numpy as np               # 수치 계산과 데이터 배열 처리에 최적화된 Numpy를 가져와 np라는 별칭으로 사용
import matplotlib as mpl         # Matplotlib 라이브러리의 전체 설정을 전역적으로 제어하기 위해 본체 모듈을 가져옴

# [설정] 그래프에서 마이너스(-) 부호가 폰트 문제로 깨져서 나오는 현상을 방지함 (True가 기본값이지만 False로 설정하여 기호 유지를 강제함)
mpl.rcParams['axes.unicode_minus'] = False

# [설정] 컴퓨터의 난수 생성 알고리즘을 0번 시드로 고정함. 이를 통해 코드를 실행할 때마다 항상 동일한 '랜덤' 결과가 나오게 하여 재현성을 확보함
np.random.seed(0)


# [함수: plt.figure] 
# 새로운 독립된 그래프 창(도화지)을 생성하는 함수.
# - figsize=(가로, 세로): 도화지의 크기를 인치(inch) 단위로 설정함.
# - f1이라는 변수에 생성된 '도화지 객체 자체'를 저장함.
f1 = plt.figure(figsize=(10, 2))

# [함수: plt.title] 
# 현재 가장 마지막에 생성되었거나 선택된 '활성 도화지'의 최상단에 제목 문자열을 출력하는 함수.
plt.title('figure size : (10, 2)')

# [함수: plt.plot / np.random.randn]
# - np.random.randn(100): 평균 0, 표준편차 1인 가우시안 표준정규분포 난수 100개를 생성함.
# - plt.plot(): 생성된 100개의 데이터를 y값으로 사용하여 선 그래프를 그림. x축은 자동으로 0부터 99까지 할당됨.
plt.plot(np.random.randn(100))

# [함수: plt.gcf] 
# 'Get Current Figure'의 약자로, 현재 메모리 상에서 Matplotlib이 
# '지금 편집 중'이라고 인식하는 도화지 객체를 찾아내서 반환함.
# 여기서는 f1 외에 다른 도화지를 만든 적이 없으므로, f2 역시 f1과 똑같은 도화지를 가리키게 됨.
f2 = plt.gcf()

# [함수: print / id]
# - f1, f2: 각 객체가 어떤 형태인지(Figure 객체 정보)를 출력함.
# - id(): 파이썬 메모리 상에서 해당 객체가 저장된 고유 주소값(주민번호와 같은 역할)을 반환함.
# - 만약 id(f1) == id(f2) 라면, 두 변수는 이름만 다를 뿐 실제로는 같은 '하나의 도화지'를 보고 있다는 증거임.
print(f1, id(f1)) # f1 객체의 타입과 메모리 고유 번호를 콘솔창에 출력
print(f2, id(f2)) # f2 객체의 타입과 메모리 고유 번호를 콘솔창에 출력

# [함수: plt.show]
# 메모리 상에 쌓여있는 모든 피규어(도화지) 객체들을 실제 윈도우 창으로 띄워서 사용자에게 보여주는 함수.
# 이 함수가 실행된 후 창을 닫아야만 다음 파이썬 코드가 진행됨.
plt.show()
```

```python
import matplotlib.pyplot as plt # 그래프 시각화를 위한 pyplot 모듈 임포트
import numpy as np               # 수치 계산 및 데이터 배열 생성을 위한 numpy 임포트
import matplotlib as mpl         # Matplotlib 전역 설정을 제어하기 위한 모듈 임포트

# [설정] 그래프 내 마이너스(-) 기호가 깨지는 현상을 방지함
mpl.rcParams['axes.unicode_minus'] = False

# [데이터 생성: np.linspace] 
# 지정된 범위(start~stop)를 num 개수만큼 균일하게 나눈 배열을 생성함
x1 = np.linspace(start=0.0, stop=5.0, num=50) # 0부터 5까지 50개의 점 생성
x2 = np.linspace(0.0, 2.0, num=50)             # 0부터 2까지 50개의 점 생성

# [수학 함수 적용]
# y1: 지수 함수(exp)와 코사인(cos)을 곱해 시간이 갈수록 진폭이 줄어드는 '감쇠 진동' 구현
y1 = np.cos(2*np.pi * x1) * np.exp(-x1)
# y2: 일정한 진폭을 유지하는 일반적인 코사인 곡선 구현
y2 = np.cos(2*np.pi * x2)

# [함수: plt.suptitle] 
# Super Title의 약자로, 여러 개의 서브플롯을 포함하는 전체 도화지의 '통합 제목'을 설정함
plt.suptitle('Main Title')

# --- 첫 번째 서브플롯 영역 시작 ---
# [함수: plt.subplot(행, 열, 인덱스)]
# 도화지를 2행 1열로 나누고, 그 중 '첫 번째(위쪽)' 칸을 작업 영역으로 선택함
plt.subplot(2, 1, 1)
# 'yo-': Yellow(노란색), 원형 마커(o), 실선(-) 스타일로 그래프를 그림
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2-1 subplots') # 첫 번째 칸 전용 제목
plt.xlabel('time (m)')              # 첫 번째 칸 x축 라벨
plt.ylabel('Damped oscillation')    # 첫 번째 칸 y축 라벨 (철자 교정: oscillation)

# --- 두 번째 서브플롯 영역 시작 ---
# [함수: plt.subplot(2, 1, 2)] 
# 도화지의 2행 1열 중 '두 번째(아래쪽)' 칸을 선택함 
# (원본 코드의 2,1,1 부분을 2,1,2로 수정해야 아래 칸에 그려집니다)
plt.subplot(2, 1, 2)
# 'r.-': Red(빨간색), 점 마커(.), 실선(-) 스타일로 그래프를 그림
plt.plot(x2, y2, 'r.-')
# [주의] plt.plot('문자열')은 에러가 날 수 있으므로 plt.title()을 사용하는 것이 맞음
plt.title('A tale of 2-2 subplots')
plt.xlabel('time (s)')              # 두 번째 칸 x축 라벨
plt.ylabel('Undamped')              # 두 번째 칸 y축 라벨

# [함수: plt.tight_layout]
# 서브플롯 간의 간격을 자동으로 조절하여 제목이나 축 라벨이 서로 겹치지 않게 배치해주는 아주 유용한 함수
plt.tight_layout()

# 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt라는 별칭으로 지정함

# [함수: plt.subplots()]
# - fig(Figure): 전체 도화지 객체를 생성함
# - ax0(Axes): 실제 그래프가 그려지는 첫 번째 좌표축 영역을 생성함
# - 이 방식은 객체지향형 코딩으로, 그래프의 각 요소를 더 세밀하게 제어할 수 있게 함
fig, ax0 = plt.subplots()

# [함수: ax0.twinx()]
# - 'Twin X-axis'의 약자로, 기존 ax0의 x축(가로축)을 그대로 공유하면서
# - y축(세로축)만 오른쪽(Right)에 새로 하나 더 만드는 함수임
# - 이제 ax0는 왼쪽 y축을, ax1은 오른쪽 y축을 담당하게 됨
ax1 = ax0.twinx()

# --- 첫 번째 y축(왼쪽, ax0) 설정 영역 ---
# [함수: set_title] 현재 도화지의 메인 제목을 설정함
ax0.set_title('2개의 y축 한 figure에서 사용하기')

# [함수: plot] 데이터를 빨간색 실선('r-')으로 그리며, 범례 이름을 'y0'로 지정함
# 데이터 값들이 10 이하의 작은 수치들임
ax0.plot([10, 5, 2, 9 ,7], 'r-', label='y0')

# [함수: set_ylabel / set_xlabel] 각 축의 이름표(Label)를 작성함
ax0.set_ylabel('y0')             # 왼쪽 세로축 이름
ax0.set_xlabel('공유되는 x축')     # 아래쪽 가로축 이름 (ax1과 같이 씀)

# [함수: grid(False)] 격자무늬(Grid)를 표시하지 않도록 설정함
ax0.grid(False)

# --- 두 번째 y축(오른쪽, ax1) 설정 영역 ---
# [함수: plot] 데이터를 초록색 점선('g:')으로 그리며, 범례 이름을 'y1'로 지정함
# 데이터 값들이 100~200 사이로, 왼쪽 데이터(y0)보다 약 20배 이상 큼
ax1.plot([100, 200, 220, 180, 120], 'g:', label='y1')

# [함수: set_ylabel] 오른쪽 세로축의 이름표를 'y1'로 설정함
ax1.set_ylabel('y1')

# [함수: grid(False)] 오른쪽 축 기준의 격자도 숨김 (왼쪽/오른쪽 격자가 겹치면 지저분해 보이기 때문)
ax1.grid(False)

# [함수: plt.show()] 메모리에 그려진 모든 설정값을 화면에 최종 출력함
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 및 데이터 생성을 위한 numpy 임포트

# [함수: plt.subplots(행, 열)]
# - 2행 2열 구조의 서브플롯(총 4칸)을 생성함
# - fig: 전체 도화지 객체
# - ax: 4개의 그래프 영역이 담긴 2차원 배열 (ax[행, 열] 형태로 접근)
fig, ax = plt.subplots(2, 2)

# --- [0, 0] 위치: 선 그래프 (Line Plot) ---
# - np.linspace(0, 100, 20): 0부터 100까지 20개의 점 생성
# - marker='o', ms=2: 원형 마커를 크기 2로 표시
# - mec='r': 마커의 테두리 색상을 빨간색(red)으로 설정
ax[0, 0].plot(np.linspace(0, 100, 20), np.linspace(0, 100, 20)**2, marker='o', markersize=2,
              markeredgecolor='r')

# --- [0, 1] 위치: 히스토그램 (Histogram) ---
# - np.random.randn(500): 500개의 정규분포 난수 생성
# - bins=30: 데이터를 30개의 구간으로 나누어 빈도를 측정
# - alpha=0.5: 막대의 투명도를 0.5(50%)로 설정하여 부드럽게 표현
ax[0, 1].hist(np.random.randn(500), bins=30, color='b', alpha=0.5)

# --- [1, 0] 위치: 막대 그래프 (Bar Chart) ---
# - [1, 3, 5, 7]: 막대가 세워질 x축의 위치
# - np.array([1, 2, 3, 4]): 각 막대의 높이(데이터 값)
# - color='r': 막대 색상을 빨간색으로 설정
ax[1, 0].bar([1, 3, 5, 7], np.array([1, 2, 3, 4]), color='r')

# --- [1, 1] 위치: 박스 플롯 (Box Plot) ---
# - np.random.randn(500, 5): 500행 5열의 난수 생성 (총 5개의 데이터 그룹)
# - 데이터의 최댓값, 최솟값, 중앙값 및 사분위수를 한눈에 보여주는 통계 그래프
ax[1, 1].boxplot(np.random.randn(500, 5))

# 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정

# 1. 딕셔너리 형태의 데이터 준비
# 'data_x'라는 키에는 [1, 2, 3, 4, 5] 리스트가,
# 'data_y'라는 키에는 [2, 3, 5, 10, 8] 리스트가 저장되어 있음
data_dict = {'data_x' : [1, 2, 3, 4, 5], 'data_y' : [2, 3, 5, 10, 8]}

# 2. plt.plot()에서 data 인자 사용하기
# - 'data_x' : x축에 사용할 데이터의 '키(Key)' 이름
# - 'data_y' : y축에 사용할 데이터의 '키(Key)' 이름
# - data=data_dict : 위에서 지정한 키 이름들을 찾을 원본 데이터 소스(딕셔너리)를 지정함
# 이 방식을 쓰면 코드가 훨씬 직관적이고 읽기 쉬워짐
plt.plot('data_x', 'data_y', data=data_dict)

# 3. 최종 그래프 출력
plt.show()
```

```python
# 산점도 및 버블 차트 예제

import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 및 난수 생성을 위해 np로 별칭 지정

# 1. 데이터 준비
N = 30                          # 생성할 점의 개수 설정
np.random.seed(0)               # 실행할 때마다 동일한 랜덤 값이 나오도록 시드 고정

x = np.random.rand(N)           # x축 좌표: 0~1 사이의 난수 30개
y1 = np.random.rand(N)          # y축 좌표: 0~1 사이의 난수 30개
y2 = np.random.rand(N)          # 색상(Color) 기준 데이터: 0~1 사이의 난수 30개
y3 = np.pi * (15 * np.random.rand(N)**2) # 크기(Size) 기준 데이터: 점의 면적을 결정함

# 2. 그래프 제목 설정
plt.title('Bubble Chart')

# 3. plt.scatter() 함수 상세 옵션 해석:
# - x, y1 : 점들이 찍힐 평면상의 좌표
# - c=y2   : color, y2 값에 따라 점의 색상을 다르게 지정함 (수치 데이터)
# - s=y3   : size, y3 값에 따라 점의 크기(면적)를 다르게 지정함
# - alpha=0.7 : 점의 투명도를 0.7로 설정 (점이 겹쳤을 때 뒤쪽 점이 보이게 함)
# - cmap='Spectral' : 수치 데이터(y2)를 어떤 색상 테마로 매핑할지 결정 (무지개색 계열)
plt.scatter(x, y1, c=y2, s=y3, alpha=0.7, cmap='Spectral') 

# 4. 컬러바(Colorbar) 추가:
# 그래프 옆에 색상이 어떤 수치(y2)를 의미하는지 나타내는 범례 막대를 표시함
plt.colorbar()

# 5. 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 및 난수 생성을 위해 np로 별칭 지정

# 1. 데이터 준비
N = 30                          # 생성할 데이터 포인트(점)의 개수를 30개로 설정
np.random.seed(0)               # 실행할 때마다 동일한 난수가 발생하도록 시드(Seed) 번호 고정

x = np.random.rand(N)           # x축 좌표: 0에서 1 사이의 균일 분포 난수 30개 생성
y1 = np.random.rand(N)          # y축 좌표: 0에서 1 사이의 균일 분포 난수 30개 생성
y2 = np.random.rand(N)          # 색상(Color) 데이터: 각 점의 색상을 결정할 0~1 사이의 난수

# 2. 버블 크기 계산 (Size)
# (15 * np.random.rand(N))**2 : 반지름을 제곱하여 면적 개념으로 크기를 만듦
# np.pi를 곱하여 원의 넓이($\pi r^2$) 공식과 유사한 형태로 버블의 크기(s) 데이터를 생성함
y3 = np.pi * (15 * np.random.rand(N))**2 

# 3. 그래프 제목 설정
plt.title("Bubble Chart")

# 4. plt.scatter() 함수 상세 옵션 해석:
# - x, y1 : 점들이 위치할 가로축과 세로축의 좌표 데이터
# - c=y2   : color, y2의 수치에 따라 점의 색상을 다르게 지정함
# - s=y3   : size, y3의 수치에 따라 점의 크기(면적)를 다르게 지정함
# - alpha=0.7 : 투명도를 0.7로 설정 (점이 겹쳤을 때 뒤에 가려진 점도 보이게 함)
# - cmap="Spectral" : 수치를 색상으로 변환할 때 사용할 색상 지도(Colormap) 테마
plt.scatter(x, y1, c=y2, s=y3, alpha=0.7, cmap="Spectral") 

# 5. 컬러바(Colorbar) 표시:
# 그래프 오른쪽에 색상이 어떤 수치(y2)를 의미하는지 보여주는 범례 막대를 추가함
plt.colorbar()

# 6. 최종 그래프 출력 (주의: plt.show 뒤에 괄호를 붙여야 실행됩니다)
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 및 그리드 생성을 위해 np로 별칭 지정

# [함수 정의: f(x, y)]
# x와 y 좌표를 입력받아 높이값(z)을 계산하는 3차원 곡면 함수임
# 지수 함수(exp)를 곱해 중심부에서 주변으로 갈수록 값이 0에 수렴하도록 설계됨
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

# 1. 데이터 준비
n = 256                         # 해상도를 결정하는 점의 개수 (많을수록 부드러운 곡선이 됨)
x = np.linspace(-3, 3, n)       # -3부터 3까지 n개의 점 생성 (x축 범위)
y = np.linspace(-3, 3, n)       # -3부터 3까지 n개의 점 생성 (y축 범위)

# [함수: np.meshgrid]
# 1차원 배열인 x와 y를 조합해 2차원 평면 위의 모든 격자점(Grid) 좌표를 만듦
# xx는 모든 점의 x좌표, yy는 모든 점의 y좌표를 담은 행렬이 됨
xx, yy = np.meshgrid(x, y)

# 2차원 격자점 전체를 함수 f에 대입하여 각 지점의 높이값(zz)을 계산함
zz = f(xx, yy)

# 2. 그래프 그리기
plt.title('Counter plots')      # 그래프 상단 제목 설정 (철자 교정: Contour plots)

# [함수: plt.contourf]
# Filled Contour의 약자로, 등고선 사이의 영역을 색상으로 채우는 함수임
# - alpha=0.75 : 색상의 투명도를 설정
# - cmap='hot' : 높낮이에 따라 따뜻한 색 계열(검정-빨강-노랑)로 색상을 입힘
plt.contourf(xx, yy, zz, alpha=0.75, cmap='hot')

# [함수: plt.contour]
# 색상 위에 실제 등고선(Line)을 그
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 및 배열 조작을 위해 np로 별칭 지정

# 1. 데이터 준비
y1 = [70, 90, 80]              # 본인의 과목별 점수 리스트
y2 = [60, 70, 90]              # 반 전체의 과목별 평균 점수 리스트
x = np.arange(len(y1))         # 기본 x축 좌표 생성 ([0, 1, 2])

# 2. 오차(Error) 데이터 생성
# np.random.randn(3)으로 난수를 발생시킨 뒤, 절대값(abs)을 취하고 5를 더해 오차 범위를 만듦
err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5

# x축에 표시될 과목명 리스트
xlabel = ['국어', '영어', '수학']

# 그래프 전체 제목 설정
plt.title('Bar chart')

# 3. 첫 번째 막대(점수) 그리기
# - x : 막대가 시작될 기본 위치
# - y1 : 막대의 높이(데이터 값)
# - yerr = err1 : 오차 막대(Error bar)의 크기를 지정함
# - ecolor = 'yellow' : 오차 막대의 선 색상을 노란색으로 설정
# - width = 0.3 : 막대의 폭을 0.3으로 좁게 설정 (옆에 다른 막대를 세우기 위함)
# - color = 'red' : 막대의 본체 색상을 빨간색으로 설정
# - align = 'center' : 지정한 x좌표를 기준으로 막대를 중앙 정렬함
# - label = '점수' : 범례에 표시될 이름 지정
plt.bar(x, y1, yerr=err1, ecolor='yellow', width=0.3, color='red', align='center', label='점수')

# 4. 두 번째 막대(반평균) 그리기
# - x + 0.4 : 첫 번째 막대(폭 0.3)와 겹치지 않도록 오른쪽으로 0.4만큼 밀어서 그림
# - y2 : 막대의 높이(데이터 값)
# - yerr = err2 : 오차 막대 크기 지정
# - width = 0.3 : 동일한 폭 설정
# - color = 'blue' : 막대의 본체 색상을 파란색으로 설정
# - label = '반평균' : 범례에 표시될 이름 지정
plt.bar(x+0.4, y2, yerr=err2, width=0.3, color='blue', align='center', label='반평균')

# 5. 축 및 디자인 설정
# - plt.xticks() : 숫자 좌표 대신 '과목명'을 출력함
#   x + 0.4/2 를 통해 두 막대 사이의 중앙 지점에 글자가 오도록 위치를 조정함
plt.xticks(x+0.4/2, xlabel)

plt.ylim(0, 100)               # y축의 범위를 0점에서 100점으로 고정
plt.xlabel('과목')              # x축 이름표
plt.ylabel('점수')              # y축 이름표

# 범례(Legend) 표시 : '점수'와 '반평균' 설명 상자를 띄움
plt.legend()

# 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 및 배열 조작을 위해 np로 별칭 지정

# 1. 데이터 준비
y1 = [70, 90, 80]              # 본인의 과목별 점수 리스트 (가로 막대의 길이가 됨)
y2 = [60, 70, 90]              # 반 전체의 과목별 평균 점수 리스트
x = np.arange(len(y1))         # 세로축(y축)에 배치될 기본 위치 좌표 ([0, 1, 2])

# 2. 오차(Error) 데이터 생성
# 데이터의 변동성을 보여주기 위한 임의의 오차값 생성
err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5

# y축(세로축)에 표시될 과목명 리스트
ylabel = ['국어', '영어', '수학']

# 그래프 전체 제목 설정
plt.title('Bar chart')

# 3. 첫 번째 가로 막대(점수) 그리기 (plt.barh)
# - x : 막대가 시작될 y축상의 위치 (눕힌 그래프이므로 첫 번째 인자가 y축 좌표임)
# - y1 : 막대의 가로 길이(데이터 값)
# - xerr = err1 : 가로 방향의 오차 막대(Error bar) 크기 지정
# - ecolor = 'yellow' : 오차 막대의 선 색상
# - height = 0.3 : 막대의 두께(세로폭)를 0.3으로 설정 (세로 막대의 width와 대응됨)
# - color = 'red' : 막대 본체 색상
# - align = 'center' : 지정한 좌표를 기준으로 막대를 중앙 정렬
# - label = '점수' : 범례 이름
plt.barh(x, y1, xerr=err1, ecolor='yellow', height=0.3, color='red', align='center', label='점수')

# 4. 두 번째 가로 막대(반평균) 그리기
# - x + 0.3 : 첫 번째 막대와 겹치지 않도록 위쪽으로 0.3만큼 밀어서 그림
# - y2 : 막대의 가로 길이
# - xerr = err2 : 가로 방향 오차 막대
# - height = 0.3 : 동일한 두께 설정
# - color = 'blue' : 막대 본체 색상
# - label = '반평균' : 범례 이름
plt.barh(x+0.3, y2, xerr=err2, height=0.3, color='blue', align='center', label='반평균')

# 5. 축 및 디자인 설정
# - plt.yticks() : 숫자 좌표 대신 '과목명'을 세로축에 출력함
#   두 막대의 중앙인 x + 0.3/2 지점에 글자를 배치하여 가독성을 높임
plt.yticks(x+0.3/2, ylabel)

plt.xlim(0, 100)               # 가로축(점수)의 범위를 0에서 100으로 고정
plt.ylabel('과목')              # 세로축 이름표
plt.xlabel('점수')              # 가로축 이름표

# 범례(Legend) 표시
plt.legend()

#
```

```python
import matplotlib.pyplot as plt

# 1. 도화지 설정
# 가로 10, 세로 5 비율로 설정하여 연도별 변화를 시원하게 표현함
plt.figure(figsize = (10,5))

# 2. 제목 및 축 레이블 설정 (한글 폰트 설정이 되어 있어야 정상 출력됨)
plt.title("학년도 기준 수능 과학탐구 응시자 수")
plt.xlabel('학년도')
plt.ylabel('응시자 수(단위 : 만명)')

# 3. 누적 막대 그래프 그리기 (층층이 쌓기)
# [1층] 지구과학: 가장 바닥(0)에서 시작함
plt.bar(sci['year'], sci['earth'], label = '지구과학')

# [2층] 생명과학: 지구과학의 높이만큼 띄워서 시작함
plt.bar(sci['year'], sci['life'], bottom = sci['earth'] ,label = '생명과학')

# [3층] 화학: 지구과학 + 생명과학의 합계 높이만큼 띄워서 시작함
plt.bar(sci['year'], sci['chemistry'], bottom = sci['earth'] + sci['life'], label = '화학')

# [4층] 물리학: 앞선 세 과목(지구과학 + 생명과학 + 화학)의 총합을 바닥으로 설정함
plt.bar(sci['year'], sci['physics'], bottom = sci['earth'] + sci['life'] + sci['chemistry'], label = '물리학')

# 4. 범례 설정
# ncol = 2 : 범례 항목이 4개이므로, 한 줄에 2개씩 2줄로 배치하여 그래프 공간을 효율적으로 사용함
plt.legend(ncol = 2)

# 5. 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정

# 1. 데이터 준비
# 각 조각의 이름(labels)과 실제 수치(sizes) 정의
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15.55, 30, 45, 9.45]

# 각 조각에 입힐 색상 리스트
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# [함수: explode] 
# 특정 조각을 원점으로부터 얼마나 떨어뜨릴지 결정 (0.1은 반지름의 10%만큼 밖으로 밀어냄)
# 여기서는 두 번째 항목인 '돼지'만 0.1만큼 튀어나오게 설정하여 강조함
explode = (0, 0.1, 0, 0)

# 2. 그래프 제목 설정 (철자 교정: Pie chart)
plt.title('Pipchar')

# 3. plt.pie() 함수 상세 옵션 해석:
# - sizes : 각 조각의 크기 데이터
# - explode : 조각 돌출 효과 적용
# - labels : 각 조각의 이름표 표시
# - colors : 설정한 색상 리스트 적용
# - autopct : 'auto percent'의 약자로, 조각 위에 백분율을 자동으로 표시함
#   - '%1.2f%%' : 소수점 아래 둘째 자리까지 표시하고 뒤에 % 기호를 붙임
# - shadow = True : 차트 아래에 그림자를 생성하여 입체감을 부여함
# - startangle = 90 : 첫 번째 조각('개구리')이 12시 방향(90도)에서 시작하도록 설정
# - counterclock = False : 시계 방향으로 조각을 배치함 (True일 경우 반시계 방향)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
        autopct='%1.2f%%', shadow=True, startangle=90, counterclock=False)

# [함수: plt.axis('equal')] 
# 파이 차트가 찌그러진 타원이 아닌 완벽한 원형을 유지하도록 가로세로 비율을 동일하게 맞춤
plt.axis('equal')

# 4. 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정

# 1. 데이터 및 색상 준비
# 각 과일 항목의 비율과 이름표 설정
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
# 헥사코드(Hex Code)를 사용해 파스텔 톤의 세련된 색상 지정
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']

# 2. 부채꼴 조각(Wedge)의 세부 속성 설정 (핵심 포인트!)
# - 'width': 0.7 : 부채꼴의 너비를 반지름의 70%로 설정함. 나머지 30%가 비워지며 '도넛 모양'이 됨.
# - 'edgecolor': 'b' : 각 조각의 테두리 색상을 파란색(blue)으로 설정함.
# - 'linewidth': 5 : 테두리 선의 두께를 5포인트로 아주 두껍게 설정하여 강조함.
wedgeprops={'width': 0.7, 'edgecolor': 'b', 'linewidth': 5}

# 3. plt.pie() 함수 상세 옵션 해석:
# - ratio, labels, colors : 데이터, 이름, 색상 적용
# - autopct='%.1f%%' : 소수점 첫째 자리까지 백분율 표시
# - startangle=260 : 그래프의 시작 지점을 260도 방향으로 회전시켜 배치함
# - counterclock=False : 시계 방향으로 조각을 배치함
# - wedgeprops=wedgeprops : 위에서 정의한 도넛 모양 및 테두리 스타일을 적용함
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, 
        counterclock=False, colors=colors, wedgeprops=wedgeprops)

# 4. 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 및 난수 생성을 위해 np로 별칭 지정

# 1. 데이터 준비
np.random.seed(0)               # 실행할 때마다 동일한 랜덤 값이 나오도록 시드 고정
# 표준정규분포(평균 0, 표준편차 1)를 따르는 난수 100개 생성
x = np.random.randn(100)

# 2. 그래프 제목 설정
plt.title('Histogram')

# 3. plt.hist() 함수 상세 옵션 해석:
# - x : 히스토그램을 그릴 대상 데이터 배열
# - bins=10 : 전체 데이터를 10개의 구간(기둥)으로 나누어 빈도를 측정함
# - rwidth=1 : 막대 간의 상대적인 너비 (1은 빈틈없이 꽉 채우는 설정)
# - color='red' : 막대의 색상을 빨간색으로 설정
# - alpha=0.4 : 투명도를 0.4로 설정하여 배경 격자나 겹치는 막대가 보이게 함
# - density=True : [중요] 막대의 높이를 '개수'가 아닌 '확률 밀도'로 표시함.
#   이 옵션을 켜면 막대들의 전체 면적 합이 1이 되어 확률 분포 함수와 비교하기 좋아짐
plt.hist(x, bins=10, rwidth=1, color='red', alpha=0.4, density=True)

# 주석 처리된 부분 예시: 구간(bins)을 20개로 늘리면 더 세밀한 분포 확인 가능
# plt.hist(x, bins=20, rwidth=1, color='blue', alpha=0.4, density=True)

# 4. 최종 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt # 시각화 라이브러리를 불러와 plt로 별칭 지정
import numpy as np               # 수치 계산 및 난수 생성을 위해 np로 별칭 지정

# 1. 데이터 준비
np.random.seed(0)               # 실행할 때마다 동일한 랜덤 값이 나오도록 시드 고정
# 표준정규분포(평균 0, 표준편차 1)를 따르는 난수 100개 생성
x = np.random.randn(100)

# 2. 그래프 제목 설정
plt.title('Histogram')

# 3. 첫 번째 히스토그램 (빨간색, 구간 10개)
# - bins=10 : 전체 데이터를 10개의 뭉치로 나누어 전체적인 흐름을 파악함
# - rwidth=1 : 막대 사이의 간격 없이 꽉 채움
# - color='red', alpha=0.4 : 빨간색이며 40%의 투명도를 가짐 (뒤의 파란색이 비치게 함)
# - density=True : 막대의 높이를 개수가 아닌 확률 밀도로 표시하여 전체 면적 합을 1로 만듦
plt.hist(x, bins=10, rwidth=1, color='red', alpha=0.4, density=True)

# 4. 두 번째 히스토그램 (파란색, 구간 20개)
# - bins=20 : 데이터를 더 세밀하게 20개의 뭉치로 나누어 굴곡을 자세히 확인함
# - color='blue', alpha=0.4 : 파란색이며 동일한 투명도를 적용함
# - density=True : 빨간색 그래프와 높이 단위를 맞추기 위해 동일하게 설정함
plt.hist(x, bins=20, rwidth=1, color='blue', alpha=0.4, density=True)

# 5. 최종 그래프 출력
plt.show()
```
