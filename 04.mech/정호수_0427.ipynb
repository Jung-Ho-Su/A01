(2) 레이블 인코딩: 기존 데이터 프레임을 새로은 데이터 프레임으로 저장한 다음 처리한 결과를 기존 데이터 컬럼은 삭제하고 새로운 데이터 프레임에 재저장
- Sex
- Embarked

```python
from sklearn.preprocessing import LabelEncoder

# 1. 기존 데이터프레임을 새로운 데이터프레임으로 복사하여 저장
df_encoded = df.copy()

# 2. 레이블 인코더 생성
le = LabelEncoder()

# 3. Sex, Embarked 처리
# 처리 결과를 새로운 컬럼으로 저장한 후 기존 컬럼 삭제
for col in ['Sex', 'Embarked']:
    new_col_name = col + '_encoded'
    df_encoded[new_col_name] = le.fit_transform(df_encoded[col])
    
    # 기존 데이터 컬럼 삭제
    df_encoded.drop(col, axis=1, inplace=True)
    
    # 처리된 결과를 다시 기존 컬럼명으로 변경 (재저장)
    df_encoded.rename(columns={new_col_name: col}, inplace=True)

# 4. 결과 확인
df_encoded.head()
```

(3) 원핫인코딩: 기존 데이터 프레임을 새로은 데이터 프레임으로 저장한 다음 처리한 결과를 기존 데이터 컬럼은 삭제하고 원핫인코더로 변환한 내용을 새로운 데이터 프레임에 추가할때 컬럼명은 적당히 정해서 추가한다.
- Sex: 
- Embarked

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

# 1. 기존 데이터프레임(df_encoded)을 새로운 데이터프레임(df_new)으로 복사
df_new = df_encoded.copy()

# 2. 원-핫 인코딩할 컬럼 정의 (데이터셋의 실제 컬럼명인 대문자로 수정)
target_cols = ['Sex', 'Embarked']

# 3. OneHotEncoder 설정
# sparse_output=False로 설정하면 .toarray()를 호출한 것과 같은 밀집 행렬을 반환합니다.
# dtype=int를 추가하여 0과 1로 나오게 설정합니다.
enc = OneHotEncoder(sparse_output=False, dtype=int)

for col in target_cols:
    # 원-핫 인코딩 적용 (2차원 구조로 입력해야 하므로 [[col]] 사용)
    encoded_data = enc.fit_transform(df_new[[col]])
    
    # 새로운 컬럼명 생성 (예: Sex_0, Sex_1 / Embarked_0, Embarked_1, Embarked_2)
    column_names = [f"{col}_{cat}" for cat in enc.categories_[0]]
    
    # 인코딩된 결과를 데이터프레임으로 변환하여 병합
    encoded_df = pd.DataFrame(encoded_data, columns=column_names, index=df_new.index)
    df_new = pd.concat([df_new, encoded_df], axis=1)
    
    # 기존 원본 컬럼 삭제
    df_new.drop(col, axis=1, inplace=True)

# 4. 결과 확인
print("변환 후 컬럼 리스트:", df_new.columns.tolist())
df_new.head()
```
