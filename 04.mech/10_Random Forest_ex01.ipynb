```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = datasets.load_iris()
print('아이리스 종류 :', iris.target_names)
print('데이터 열 이름 :', iris.feature_names)
```

```python
data = pd.DataFrame(
    {
        'sepla_length': iris.data[:, 0],
        'sepla_width': iris.data[:, 1],
        'petal_length': iris.data[:, 2],
        'petal_width': iris.data[:, 3],
        'species': iris.target
    }
)
data.head(2)
```

```python
# 데이터 분리 시키는 작업
```

```python
# 데이터 분리시키는 작업
x = data[['sepla_length', 'sepla_width', 'petal_length', 'petal_width']]
y = data['species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
x_train.shape, y_train.shape
```

```python
# 모델 설계하는 부분
```

### 랜덤포레스트 하이퍼 파라미터
- n_estimators : 결정트리의 갯수를 지정, default = 100
- max_features : 최적의 ㅂㄴ할을 위해 고려할 최데 feature 개수, default = 'sqrt', {'sqrt', 'log2', None}
- booststrap : 부트스트랩(중복허용 샘플링) 사용 여부, default = True # 부트 스트랩을 사용하겠다는 의미

```python
forest = RandomForestClassifier(n_estimators=100)
forest
```

```python
# 나머지는 의사결정트리와 똑같아서 여기에 정리 안한다고 함...ㅠㅠ
```

```python
# 훈련
```

```python
forest.fit(x_train, y_train)
```

```python
# 분류/예측
y_pred = forest.predict(x_test)
print(y_pred)

print(list(y_pred))
```

```python
#  스코어를 계산하는 방법
```

```python
print("정확도 :", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_importances_values = forest.feature_importances_
feature_importances = pd.Series(feature_importances_values, index = x_train.columns)
feature_top4 = feature_importances.sort_values(ascending=False)[:4]

plt.figure(figsize=(8,6))
plt.title('Top 4 Feature Importances')
sns.barplot(x=feature_top4, y=feature_top4.index)
plt.show()
```

```python
# 현재 분류를 했는데 얘를 회귀모델에도 사용할 수 있다.
```

### RandomForestRegressor

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('boston_house.csv')
df.head(2)
```

```python
X = df.drop('Target', axis=1)
y = df['Target']
```

```python
# 모델 설계


reg = RandomForestRegressor(
    n_estimators=50,                  # 우리가 만들고 싶어하는 dt 개수 
    criterion='squared_error',        # 파라매터는 그냥 트리와 같은데 변수가 다르다는것을 알고 있을 것
    max_depth=4,
    max_features='sqrt',
    max_samples=1.0,
    bootstrap=True,
    oob_score=True,                   
    random_state=100).fit(X,y)
```

```python
## bootstrap=True, oob_score=True,

# 기본값을 false 이며 일반화 점수를 추정하기 위해 아웃오브백 샘플을 사용할지 여부 
# 부트스트랩을 지정하게 되면 중복을 허용한 샘플링르 하는데 모든 데이터를 모두 사용하지 안아 일부가 샘플링 되고 
#일부는 남는 데이터가 되는데 이때 남는데이터를 oob 데이터라고 한다.
# 이때 이 oob의 데이터를 사용해서 성능평가를 하겠다는 의미이다.
```

```python
print(reg.oob_score_)
print('R2 : ', reg.score(X,y))
```

```python
print(reg.feature_importances_)
```
