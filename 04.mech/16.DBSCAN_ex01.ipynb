```python
# import 하는 과정
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# data를 불러오는 과정
iris = load_iris()
iris_data = iris.data
iris_data_pd = pd.DataFrame(iris_data, columns=iris.feature_names)
iris_data_pd
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.5, min_samples=10) #eps=0.5, min_samples=5 가 디폴트 값
db.fit(iris_data_pd.iloc[:, 2:4])
y_pred = db.fit_predict(iris_data_pd.iloc[:, 2:4])
plt.scatter(iris_data_pd.iloc[:, 2], iris_data_pd.iloc[:, 3], c=y_pred)
plt.title('Density Clustering')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

```python
db.labels_ # 클러스터의 정보를 확인하는 방법

# 클러스터 넘버링을 확인해봤는데 -1의 값이 있었다. 이 -1의 의미는 noise data를 의미한다. 그래서 이 데이터를 표현해보니
# 위에 plt 에서처럼 별도로 나타나게 된다는 걸 확인 할 수 있었다.

# min_samples 수를 조정하게 되면 달라질 수 있으니 가장 잘 분류하는 수를 찾는 것이 우리의 역할이다.
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.5, min_samples=5) #eps=0.5, min_samples=5 가 디폴트 값
db.fit(iris_data_pd.iloc[:, 2:4])
y_pred = db.fit_predict(iris_data_pd.iloc[:, 2:4])
plt.scatter(iris_data_pd.iloc[:, 2], iris_data_pd.iloc[:, 3], c=y_pred)
plt.title('Density Clustering')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

```python
x, cnt = np.unique(people.target, return_counts=True)
print(x)
print(cnt)
```

```python
mask = np.zeros(people.target.shape, dtype=np.bool_)
print(mask.shape)
mask[0:100]
```

```python
for target in np.unique(people.target):
    print(target)
    mask[np.where(people.target == target)[0][:50]] = 1 # 이거의 의미...?
    print(np.where(people.target == target))
```

```python
mask
```

```python
X_people = people.data[mask]
y_people = people.target[mask]
```

```python
X_people.shape, y_people.shape
```

```python
x, cnt = np.unique(y_people, return_counts=True)
print(x)
print(cnt)
```

```python
X_people = X_people / 255. # minmanxscaler
```

```python
X_people.shape
```

```python
pca = PCA(n_components=100, whiten=True, random_state=0)
pca.fit(X_people)
X_pca = pca.transform(X_people) # 바뀐 이 데이터를 가지고 학습을 진행할것이다.
print(X_pca.shape)
```

```python
# 피쳐가 줄어드면 어떻게 될까???

# 데이터 사이즈가 줄어들기 때문에 연산하는 시간이 줄어든다. > 메모리 사용량도 줄어들게 된다.
# 그럼 문제점은 없을까?
# 데이터 손실이 생기게 된다.
# 데이터가 가지고 있는 특징을 최대한 보존하면서 변환하기 때문에 데이터의 손실을 최소화하게 한다.
```

```python
# 그럼 데이터의 특징은 뭔가?
# 데이터가 현재 가지고 있는 상태 정보를 의미한다. 그럼 이 현재 상태정보를 무엇으로 표현할 수 있는가?
# 다른 데이터 정보와 위치정보이고 이 위치정보는 분산의 개념으로 표현할 수 있다.
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN()
labels =dbscan.fit_predict(X_pca)
print('고유한 레이블 \n{}'.format(np.unique(labels)))
```

```python
dbscan = DBSCAN(min_samples=3)
labels =dbscan.fit_predict(X_pca)
print('고유한 레이블 \n{}'.format(np.unique(labels)))
```

```python
dbscan = DBSCAN(min_samples=3, eps=10)
labels =dbscan.fit_predict(X_pca)
print('고유한 레이블 \n{}'.format(np.unique(labels)))
```

```python
dbscan = DBSCAN(min_samples=3, eps=15)
labels =dbscan.fit_predict(X_pca)
print('고유한 레이블 \n{}'.format(np.unique(labels)))
```

```python
print("클러스터 ㅂㄹ 포인트 수: {}".format(np.bincount(labels + 1)))
```

```python
import matplotlib.pyplot as plt

noise = X_people[labels == -1]
fig, axes = plt.subplots(4, 9, subplot_kw={'xticks':(), 'yticks':()}, figsize=(12, 4))
for image, ax in zip(noise, axes.ravel()):
    ax.imshow(image.reshape(image_shape))
```

```python
for eps in [1, 3, 5, 7, 9, 11, 13]:
    for min_samples in [3, 5, 7, 9]:
        print("\neps={}, min_samples={}".format(eps, min_samples))
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_pca)
        print("클러스터 값{}".format(np.unique(labels)))
        print("클러스터 수{}".format(len(np.unique(labels))))
        print("클러스터 크기{}".format(np.unique(labels + 1)))
```

```python
dbscan = DBSCAN(min_samples=3, eps=7)
labels = dbscan.fit_predict(X_pca)

for cluster in range(max(labels) + 1):
    mask = labels == cluster
    n_images = np.sum(mask)
    fig, axes = plt.subplots(1, n_images, figsize=(n_images * 1.5, 4), subplot_kw={'xticks':(), 'yticks':()})
    for image, label, ax in zip(X_people[mask], y_people[mask], axes):
        ax.imshow(image.reshape(image_shape))
        ax.set_title(people.target_names[label].split()[-1])
```
