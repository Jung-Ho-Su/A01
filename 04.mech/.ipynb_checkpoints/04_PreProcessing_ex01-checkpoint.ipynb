```python
import seaborn as sns

iris= sns.load_dataset("iris")
iris.head(2)
```

```python
x = iris.iloc[:,:-1]  # 전체 행과, 처음부터 마지막 열 바로 전까지 가져온다.
x.head(2) # x의 변수에 전체 행을 넣었지만 head(2)라는 함수로 2개의 값만 지정해서 2개만 보여준다
```

# 표준화

```python
from sklearn.preprocessing import scale

x_scaled = scale(x) # x값에 담긴 값을 scale()의 함수 안에 넣어서 표준화를 하고 x_scaled의 변수 안에 넣는다.
x_scaled[:5,:] # 이후 상위 5개의 행과 모든 열을 
```

```python
x_scaled.mean(axis=0)
```

```python
for scaled_mean in x_scaled.mean(axis=0):
    print("{:10.9f}".format(scaled_mean))

# 완전 0은 아니지만 0에 근사하는 10의 -16승이기 때문에 0이라고 본다.
```

```python
from sklearn.preprocessing import robust_scale
# 중앙값과 사분위수를 사용하여 데이터를 스케일링(표준화)한다.

iris_robust_scaled = robust_scale(x)
iris_robust_scaled[:5,:]
```

```python
from sklearn.preprocessing import minmax_scale
# 0과 1 사이의 범위로 압축하는 스케일링(표준화) 방식으로 가장 작은 값은 0, 가장 큰 값은 1이 된다.

iris_minmax_scale = minmax_scale(x)
iris_minmax_scale[:5,:]
```

```python
from sklearn.preprocessing import maxabs_scale
# 데이터 각 열에서 가장 큰 절대값을 찾아 모든 데이터를 그 값으로 나눕니다.

iris_maxabs_scale = maxabs_scale(x)
iris_maxabs_scale[:5,:]
```

- scale / 평균 0, 표준편차 1 / 주요 용도 : 가장 일반적임 (회귀, 로지스틱 회귀 등)
- robust_scale / 중앙값, IQR 사용 / 주요 용도 : 이상치가 많은 데이터일 때 유용
- minmax_scale / 모든 데이터를 0 ~ 1 사이로 변환(압축) / 주요 용도 : 데이터의 최대&최소가 명확할 때 (이미지 처리)
- maxabs_scale / 절대값의 최대치가 1이 되도록 변환 / 희소데이터(sparse data)나 데이터의 0(중심)을 유지해야 할 때

# 표준화 standardscaler

```python
#StandardScaler 예제
import seaborn as sns

iris = sns.load_dataset("iris")
iris.iloc[:5,:-1]
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler() # 표준화를 수행할 스케일러 객체를 생성, 이후 데이터 평균과 표준편차를 기억함
iris_scaled = sc.fit_transform(iris.iloc[:,:-1])
iris_scaled[:5,:]
```

```python
## 위에 코드하고 스케일이 같은 결과가 나온다
## 이 셀의 함수는 위의 스케일 함수를 이용해 작성한 것을 그대로 가져온것 단순히
## 바로 위의 셀인 standardscaler 의 코드와 값이 같음을 비교하고 이해하기 위해 넣은 것
from sklearn.preprocessing import scale

x_scaled = scale(x)
x_scaled[:5,:]
```

```python
iris_origin = sc.inverse_transform(iris_scaled)
iris_origin[:5,:]
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
iris_scaled = sc.fit_transform(iris.iloc[:,:-1])
iris_scaled[:5,:]
```

## 인코딩

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

iris = sns.load_dataset("iris")
y = iris.species
y

# 현재 종을 확인해보니 문자로 되어있어서 숫자로 변환하는 작업을 해야한다.
# 그렇기에 이제 머신러닝을 하기 위해서는 숫자로 된 데이터를 넣어야한다 이는
# 연산을 해야하기 때문이다. 그렇기에

# 먼저 인코딩 작업을 통해 문자를 숫자로 변환시켜준다.
```

```python
le = LabelEncoder()
le.fit(y)
species = le.transform(y)
species
```

```python
le.classes_

# 현재 문자를 숫자로 변환했는데 어떤 문자가 0 1 2 ... 변환되었는지 모르기에 
# classes_ 사용해서 어떤 문자가 0 1 2 인지 순서대로 적혀있는걸 확인하는 코드이다
```

```python
le.inverse_transform(species)

# 숫자로 변경된 값을 다시 문자로 변경하기 위해서는
# inver_transform 이라는 변수를 이용해서 다시 문자로 바꿔준다.
```

# 결측값 처리

```python
import seaborn as sns
iris = sns.load_dataset("iris")
x = iris.iloc[:,:-1]
x

# 현재 데이터에 누락된 뎅티ㅓ가 없어서 가상으로 누락된 데이터를 만들거다
```

```python
import random

# 150개 중 10개의 행 위치에다가 랜덤으로 nan을 대입하겠다는 의미
# seed 값이 없기 때문에 nana이 계속 바뀌게 될것이다.

for col in range(4):
    x.iloc[[random.sample(range(len(iris)), 10)],col] = float('nan')
x.head()
```

```python
x.mean(axis=0)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit_transform(x)[0:5,:]
```

```python
x.median(axis=0)
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='median')
imp_mean.fit_transform(x)[0:5,:]
```

# 단순 데이터 처리

```python
import seaborn as sns
from sklearn.model_selection import train_test_split

iris = sns.load_dataset("iris")

# test_size를 생략하는 경우 = 0.25로 된다.
# random_state=1 // 이게 시드값을 준것과 같은 효과를 준다.

x_train, x_test, y_train, y_test = train_test_split(iris.iloc[:,0:4], iris.species, test_size=0.3, shuffle=True, random_state=1)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```
