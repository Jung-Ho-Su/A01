```python
# seaborn이 제공하는 타이타닉 데이터에서 sex 컬럼에 대한 데이터를 레이블 인코딩으로 변환시키시오.
```

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

titanic = sns.load_dataset("titanic")
titanic.head(2)

y = titanic.sex
y
```

```python
le = LabelEncoder()
le.fit(y)
sex = le.transform(y)
sex
```

```python
le.classes_
```

```python
le.inverse_transform(sex)
```
